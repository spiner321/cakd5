- size를 활용한 텐서 사이즈 보기

- shape를 활용한 텐서 사이즈 보기

- dim을 활용한 텐서 차원 확인

- 0 차원 사이즈 확인

- 1 차원 사이즈 확인

```python
x= torch.FloatTensor([
                      [1,2],
                     [3,4],
                      [5,6],
                      [7,8]])

print(x)

#size 를 활용한 텐서 사이즈 보기
print(x.size())

#  shape 를 활용한 텐서 사이즈 보기
print(x.shape)

# dim을 활용한 텐서 차원확인
print(x.dim)

#0차원 사이즈 확인
print(x.size(0))

#1차원 사이즈 확인
print(x.shape[1])
```
