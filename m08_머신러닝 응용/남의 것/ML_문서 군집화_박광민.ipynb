```python
import pandas as pd
import glob
import os
import numpy as np
path = r'H:\내 드라이브\cakd5\workspace\m8_머신러닝응용\dataset\OpinosisDataset1.0\topics'

all_files = glob.glob(os.path.join(path, "*.data"))
filename_list = []
opinion_text = []

for file_ in all_files:
    df = pd.read_table(file_, index_col=None, header=0, encoding='latin1')
    filename_ = file_.split('\\')[-1]
    filename = filename_.split('.')[0]

    filename_list.append(filename)
    opinion_text.append(df.to_string())

document_df = pd.DataFrame(
    {'filename': filename_list, 'opinion_text': opinion_text})

document_df = document_df.sort_values('filename')

document_df.index = np.arange(0, 51)
document_df.head()
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
import string

remove_punct_dict = dict((ord(punct), None)for punct in string.punctuation)
lemma = WordNetLemmatizer()

def LemTokens(tokens):
    return [lemma.lemmatize(token) for token in tokens]

def LemNormalize(text):
    return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(tokenizer=LemNormalize, stop_words='english', ngram_range=(
    1, 2), min_df=0.05, max_df=0.85)
feature_vect = tfidf_vect.fit_transform(document_df['opinion_text'])
```

```python
from sklearn.cluster import KMeans

km_cluster = KMeans(n_clusters=5, max_iter=10000, random_state=0)
km_cluster.fit(feature_vect)
cluster_label = km_cluster.labels_
clister_centers = km_cluster.cluster_centers_
```

```python
document_df['cluster_label'] = cluster_label
document_df.head()
```

# 군집별 핵심 단어 추출

```python
cluster_centers = km_cluster.cluster_centers_

print('cluster_centers shape :', cluster_centers.shape)
print(cluster_centers)
```

```python
def get_cluster_details(cluster_model, cluster_data, cluster_nums,
                        feature_names, top_n_features=10):
    cluster_details = {}

    center_info = cluster_model.cluster_centers_
    center_descend_ind = center_info.argsort()[:, ::-1]
    #내림차순
    for i in range(cluster_nums):

        cluster_details[i] = {}

        filenames = cluster_data[cluster_data["cluster_label"]
                                 == i]["filename"]
        filenames = filenames.values.tolist()

        top_feature_values = center_info[i, :top_n_features].tolist()

        top_feature_indexes = center_descend_ind[i, :top_n_features]
        top_features = [feature_names[ind] for ind in top_feature_indexes]

        cluster_details[i]["cluster"] = i
        cluster_details[i]["top_features"] = top_features
        cluster_details[i]["top_feature_values"] = top_feature_values
        cluster_details[i]["filenames"] = filenames

    return cluster_details
```

```python
feature_names = tfidf_vect.get_feature_names()

cluster_details = get_cluster_details(cluster_model=km_cluster, cluster_data=document_df, cluster_nums=3,
                                      feature_names=feature_names, top_n_features=10)
```
