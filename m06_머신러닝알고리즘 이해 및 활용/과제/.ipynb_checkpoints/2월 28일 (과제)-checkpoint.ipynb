#### [과제] t_df의 Age 컬럼을 이용하여 새로운 컬럼 Age_cat 사용자 함수를을 이용하여 출력하세요(7개 카테고리)
---

```python
def Age_cat(x):
    if x < 10: return '어린이'
    elif x < 20: return '10대'
    elif x < 30 : return '20대'
    elif x < 40 : return '30대'
    elif x < 50 : return '40대'
    elif x < 60 : return '50대'
    elif x >= 60 : return '노인'
    
t_df['Age_cat'] = t_df['Age'].apply(Age_cat)
t_df
```

#### [과제] t_df에 대하여 결손값 처리, 불필요한 컬럼 삭제, 인코딩 등을 수행하는 전처리 과정을 하나의 사용자 함수(def transform_features(df))로 작성하여 분석용 데이터셋을 만드세요.
---

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

t_df1 = t_df.copy()

display(t_df1)
t_df1.info()
```

```python
def transform_features(df):
    df1 = df.copy()
    df1['Age'].fillna(df1['Age'].mean(), inplace = True)
    df1['Age_cat'] = df1['Age'].apply(Age_cat) 
    df1['Age_cat'] = LabelEncoder().fit_transform(df1['Age_cat']).reshape(-1, 1)
    df1['Family'] = df1['SibSp'] + df1['Parch']
    df1['Embarked'].ffill(inplace = True)
    df1.drop(['PassengerId', 'Name', 'Age', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Fare', 'Name_len'], axis = 1, inplace = True)
    return pd.get_dummies(df1)
```

```python
t_df1_en = transform_features(t_df1)
t_df1_en.head()
```

##### [과제] t_df Name 컬럼에서 Title(칭호)를 추출해서 이에 기반한 age 평규값을 이용하여 age null 처리를 한 후 개선된 부분을 설명하세요.
---

```python
t_df2 = t_df.copy()
```

```python
import re

title = []
for i in t_df2['Name']:
    a = re.search(', [\w\s]+', i).group()
    b = re.sub(', ', '', a)
    title.append(b)

t_df2['Title'] = title
titlemean = t_df2.groupby('Title').mean()['Age']
```

```python
a = dict(list(t_df2.groupby('Title')))
a['Master']
```

```python
for i in titlemean.index:
    t_df2.loc[(t_df2['Age'].isnull()) & (t_df2['Title'] == i), 'Age'] = titlemean.loc[i]['Age']
t_df2.drop('Title', axis = 1, inplace = True)
t_df2_en = transform_features(t_df2)
```
