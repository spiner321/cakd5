```python
def factorial(n):
    output = 1
    for i in range(1, n + 1):
        output *= i
#         print(i)
#         print()
        print(output)
    return output

factorial(4)
        
```

```python
# 재귀란 '자기자신을 호출하는 것'을 의미한다.
# 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를 작성하고 1!, 5!, 10!을 구하세요.
# factorial(n) - n * factorial(n-1) (n >=1)

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
     
print(factorial(3))
```

```python
# 원의 면적과 둘레의 길이를 구하는사용자 함수를 작성하여 반지름이 와 10일때 면적과 둘레를 구하세요.

def r(n):
    r2 = round(n * 2 * 3.14, 2)
    s = round(n * n * 3.14, 2)
    return r2, s

r(5)
```

```python
#

def getTextFreq(filename):
    with open(filename, 'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] += 1
            else:
                fa[c] = 1
        return fa

ret = getTextFreq('data/mydata.txt')
ret = sorted(ret.items(), key = lambda x: x[1], reverse = True)
print(ret)
```

```python
# [] -> 177회 나타남
# [e] -> [97]회 나타남

for c, f in ret:
    if c == '\n':
        continue
    print(f'[{c}] -> {[f]}회 나타남')
```

```python
# age 변수를 10, 20, 30, 40, 50, 60이상 범주로 구분하여 age_cat 변수를 생성하세요.
# feature engineering

tt.age

def age_cat(age):
    cat = ''
    if age < 20: cat = '10'
    elif age < 30: cat = '20'
    elif age < 40: cat = '30'
    elif age < 50: cat = '40'
    elif age < 60: cat = '50'
    else: cat = '60'
    return cat
tt['age_cat'] = tt['age'].apply(lambda x: age_cat(x))
tt.age_cat
tt[['age', 'age_cat']].head()
```
