```python
import random
import datetime as dt

class Recommend:
    def printRecommend(self, movie_list):    
        recommend = random.choice(movie_list)
        r_name = recommend['name']
        r_time = recommend['runtime']
        next_t = t + dt.timedelta(minutes = r_time)
        print()
        print('='*50)
        print(f'오늘의 추천 영화는 <{r_name}> 이며, 상영시간은 {r_time}분 입니다.')
        print(f'지금부터 영화를 보면 {next_t.hour}시 {next_t.minute}분에 시청이 종료됩니다.')
        print('='*50)
    def printNoRecommend(self):
        print()
        print('='*50)
        print('조건에 해당하는 작품이 없습니다. 죄송합니다.')
        print('='*50)

r = Recommend()

while 1: #  사용자 입력 받기 반복
    exe = input(
'''
***영화 추천 프로그램을 시작하시겠습니까?'[ y | n ]

>> ''')

    if exe == 'y': 
        try:

            runtime_input = int(input(
'''
* 상영시간이 몇 분 이하였으면 좋겠나요? ex) 120

>> '''))
            genre_input = int(input(
'''
* 영화 장르를 선택해주세요 

    (0) 아무거나
    (1) 코미디
    (2) 로맨스/멜로
    (3) 느와르
    (4) 공포/스릴러
    (5) SF/판타지
    (6) 음악/뮤지컬

>> '''))

            movie_list = []
            t = dt.datetime.now()

            if genre_input == 0:
                for m in movies:
                    if m['runtime'] <= runtime_input:
                        movie_list.append(m)     
                    else: pass
                if movie_list:
                    r.printRecommend(movie_list)
                else: 
                    r.printNoRecommend()

            elif genre_input in range(1,7):
                for m in movies:
                    if m['genre'] == genre_input and m['runtime'] <= runtime_input:
                        movie_list.append(m)
                    else: pass
                if movie_list:
                    r.printRecommend(movie_list)
                else: 
                    r.printNoRecommend()
            else:
                print('번호 0~6 중에서 입력해주세요')

        except ValueError: 
            print('숫자를 입력해주세요.')

        except KeyboardInterrupt: 
            print('프로그램이 비정상적으로 중단되었습니다.')

    elif exe == 'n':
        print('프로그램을 종료합니다.')
        break
        
    else: print('y 또는 n을 입력해주세요.')
```
