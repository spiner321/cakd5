```python
while 1:
    print("""내 아이의 건강을 위한 맞춤 솔루션

●아이의 연령대를 골라주세요.(영아기:1, 유아기:2, 초등학생:3, 프로그램 종료:0, 그 외의 번호를 입력하면 다시 시작합니다.')\n""")
    year = int(input('아이의 연령대를 골라주세요. 영아기:1, 유아기:2, 초등학생:3, 프로그램 종료:0, 그 외의 번호를 입력하면 다시 시작합니다. '))
    if int(year) == 1:
        while 1:
            year_choice = input('\n 자녀가 영아기입니까? (맞으면:y 이외는 취소입니다.)\n')
            if year_choice == 'y':
                elementary_num = int(input('\n 아이의 키는 얼마입니까? (영유아기의 평균 신장은 50 ~ 86cm입니다.))\n'))
                if elementary_num >= 50 and elementary_num <= 60 :
                    print(f'아이에게 맞는 영양제품은 울트락토 입니다.')
                    print('')
                    break
                elif elementary_num>= 61 and elementary_num <= 70 :
                    print(f'아이에게 맞는 영양 제품은 뉴트키즈 정크시럽 입니다.')
                    print('')
                    break
                elif elementary_num>= 71 and elementary_num <= 80 :
                    print(f'아이에게 맞는 영양 제품은 뉴트키즈 철분 입니다.')
                    print('')
                    break
                elif 81 <= elementary_num and elementary_num <= 86 :
                    print(f'아이에게 맞는 영양 제품은 징거멀즈 입니다.')
                    print('')
                    break
                else:
                    print(f'영유아기의 평균 신장인(50 ~ 86cm)의 숫자만 입력하세요. 처음부터 다시 시작합니다.')
                    print('')
                    break
            else:
                print('처음부터 다시 시도해 주세요.')
                print('감사합니다.')
                print('')
                break
        else:
            print('처음부터 다시 시도해 주세요.')
            print('감사합니다.')
            print('')
            break
        
    if int(year) == 2:
        while 1:
            year_choice = input('\n 자녀가 아동기입니까? (맞으면:y 이외는 취소입니다.)\n')
            if year_choice == 'y':
                elementary_num = int(input('\n 아이의 키는 얼마입니까? (아동기의 평균 신장은 87 ~ 120cm입니다)\n'))
                if  elementary_num  >= 87 and elementary_num <= 93 :
                    print(f'아이에게 맞는 영양제품은 스마티 팬츠 입니다.')
                    print('')
                    break
                elif elementary_num >= 94 and elementary_num <= 100 :
                    print(f'아이에게 맞는 영양 제품은 함소아 칼슘가득 입니다.')
                    print('')
                    break
                elif elementary_num >= 101 and elementary_num <= 107 :
                    print(f'아이에게 맞는 영양 제품은 칼슘비타 디 입니다.')
                    print('')
                    break
                elif elementary_num >= 108  and elementary_num <=120 :
                    print(f'아이에게 맞는 영양 제품은 키즈텐042 입니다.')
                    print('')
                    break
                else:
                    print(f'아동기의 평균 신장인(87 ~ 120cm)의 숫자만 입력하세요. 처음부터 다시 시작합니다.')
                    print('')
                    break
            else:
                print('처음부터 다시 시도해 주세요.')
                print('감사합니다.')
                print('')
                break
        else:
            print('처음부터 다시 시도해 주세요.')
            print('감사합니다.')
            print('')
            break
        
    if int(year) == 3:
        while 1:
            year_choice = input('\n 자녀가 초등학생입니까? (맞으면:y 이외는 취소입니다.)\n')
            if year_choice == 'y':
                elementary_num = int(input('\n 아이의 키는 얼마입니까? (초등학생의 평균 신장은 120 ~ 156cm입니다)\n'))
                if  elementary_num >= 120 and elementary_num <= 127 :
                    print(f'아이에게 맞는 영양제품은 하이키업 입니다.')
                    print('')
                    break
                elif elementary_num>= 128 and elementary_num <= 135 :
                    print(f'아이에게 맞는 영양 제품은 아이클타임 입니다.')
                    print('')
                    break
                elif elementary_num>= 136 and elementary_num <= 146 :
                    print(f'아이에게 맞는 영양 제품은 아이커 입니다.')
                    print('')
                    break
                elif elementary_num>= 147 and elementary_num <= 156 :
                    print(f'아이에게 맞는 영양 제품은 키움정 입니다.')
                    print('')
                    break
                else:
                    print(f'초등학생의 평균 신장인(120 ~ 156cm)의 숫자만 입력하세요. 처음부터 다시 시작합니다.')
                    print('')
                    break
            else:
                print('처음부터 다시 시도해 주세요.')
                print('감사합니다.')
                print('')
                break
        else:
            print('감사합니다.')
            print('')
            break
            
    if int(year) == 0:
        print('\n 솔루션을 종료합니다 감사합니다.\n')
        print('')
        break
```
