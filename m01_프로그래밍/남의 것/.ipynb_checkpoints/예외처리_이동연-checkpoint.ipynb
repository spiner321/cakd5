```python
# 시간 함수를 이용해서 현재 시간을 입력받아 클래스로 만들어 사용

inf = {'소':4000, '중':7000, '대':10000, '1시간':6000, '10분':1500}

print(f"""
          - 입 장 료 -

        소 형  :  {inf['소']}원
        중 형  :  {inf['중']}원
        대 형  : {inf['대']}원
     
       - 시간별 추가 금액 -
       
        1시간  :  {inf['1시간']}원
        10분   :  {inf['10분']}원
          

""")
from datetime import datetime


num = input('차량 번호를 입력하세요 ')

while 1:
    try:
        size = input('해당하는 입장료를 선택하세요 [소/중/대] ')

        if size in inf:
            time_in = datetime.now()
            h = time_in.hour
            m = time_in.minute
            print(f"""
        ==================================
                  - 주 차 권 -

              차량 번호 : {num}
              입고 시각 : {h}시 {m}분
              입 장 료  : {inf[size]}원
        
        ===================================    

        """)
            break
        else:
            raise ValueError
    
    except ValueError:
        print('[소/중/대]중에서 선택하세요')

while 1:
    try:
        out = input('출고 하시겠습니까? [네/아니요]')
    
        if out=='네':
            time = input('출고 날짜와 시각을 입력하세요 [ex)2022010312300] ')
            time_out = datetime.strptime(time, "%Y%m%d%H%M%S")
            time_diff = time_out - time_in
            H = time_diff.seconds // 3600
            M = ((time_diff.seconds // 60) - (H * 60)) // 10
        
            price = (H * 6000) + (M * 1500) + inf[size]
            break
            
        elif out=='아니요': print('출고하실때 다시 와주세요')
        else:
            raise ValueError
    except ValueError:
        print('[네/아니요]중에서 선택하세요')
    except ValueError:
        print('출고 날짜와 시각을 정확히 입력하세요')

print(f"""
        ==================================
                  - 계 산 서  -

              차량 번호 : {num}
              이용 시간 : {H}시간 {M*10}분
              이 용 료  : {price}원
        
           감사합니다! 안전운전 하세요~
        ===================================    

        """)
```
