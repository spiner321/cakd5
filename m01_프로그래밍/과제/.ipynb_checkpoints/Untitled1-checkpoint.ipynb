```python
ticket = 5

seat_a = ['a' + f'{i}' for i in range(1, 31)]
seat_b = ['b' + f'{i}' for i in range(1, 31)]
seat_c = ['c' + f'{i}' for i in range(1, 31)]
seat_d = ['d' + f'{i}' for i in range(1, 31)]
seat_e = ['e' + f'{i}' for i in range(1, 31)]
seat_f = ['f' + f'{i}' for i in range(1, 31)]
seat_g = ['g' + f'{i}' for i in range(1, 31)]
seat_h = ['h' + f'{i}' for i in range(1, 31)]
seat_i = ['i' + f'{i}' for i in range(1, 31)]
seat_j = ['j' + f'{i}' for i in range(1, 31)]
seat_sum = seat_a + seat_b + seat_c + seat_d + seat_e + seat_f + seat_g + seat_h + seat_i + seat_j

try:
    while 1:
        seat = input('좌석을 선택해 주세요 : ')

        if seat in seat_sum:

            seat_sum.remove(seat)
            ticket -= 1

            if ticket == 0:
                break
        else:
            print('다시 입력해주세요.')
except:
    print('다시 입력해주세요.')
    
```
