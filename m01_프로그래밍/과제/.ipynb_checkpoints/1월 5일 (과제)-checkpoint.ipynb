```python
# # [과제] 지유 주제로 프로그래밍을 수행하세요. 단 예외처리 3건 이상

# 영화표 끊기

# 상영관
# 영화목록
# 시간목록
# 좌석목록
# 예매 내역 확인

room_list = ['IMAX관', '4dx관', '2D 일반1관', '2D 일반2관', 'SKYBOX관']
movie_list = ['반지의 제왕 3 : 왕의 귀환', '스파이더맨 2', '해리포터 : 마법사의 돌', '트루먼 쇼', '라이온킹']
time_list = ['07:00', '10:00', '13:00', '16:00', '19:00', '21:00', '24:00']
            

print("====== CGBOX 영화관 입니다. =======")

print('''
        <현재 상영관>
        
         1. IMAX관
         2. 4dx관
         3. 2D 일반1관
         4. 2D 일반2관
         5. SKYBOX관
         ''')
while 1:
    try:
        room = int(input('상영관을 선택해주세요 (숫자 입력) : '))
        if 1 <= room <= 5:
            break
        else:
            print('\n다시 입력해주세요.')            
    except:
        print('\n다시 입력해주세요.')

print('''
     <상영 목록>

1. 반지의 제왕 3 : 왕의 귀환
2. 스파이더맨 2
3. 쇼생크 탈출
4. 해리포터 : 마법사의 돌
5. 트루먼 쇼
6. 라이온킹
     ''')
while 1:
    try:
        movie = int(input('영화를 선택해주세요 (숫자 입력) : '))
        if 1 <= movie <= 6:
            break
        else:
            print('\n다시 입력해주세요.')
    except:
        print('\n다시 입력해주세요.')     
        
while 1:
    try:
        ticket = int(input('\n영화의 인원 수를 선택해주세요 : '))
        if ticket >= 1:
            break
        else:
            print('\n다시 입력해주세요.')
    except:
        print('\n다시 입력해주세요.')     
    
print('''
    <상영 시간>

     1. 07:00
     2. 10:00
     3. 13:00
     4. 16:00
     5. 19:00
     6. 21:00
     7. 24:00
     ''')    
while 1:
    try:
        time = int(input('상영시간을 선택해주세요 (숫자 입력) : '))
        if 1 <= movie <= 7:
            break
        else:
            print('\n다시 입력해주세요.')
    except:
        print('\n다시 입력해주세요.')     
           
print('''
<좌석은 a~j와 1~30을 조합하여 직접입력 (예 : a2)>
''')           
while 1:
    try:
        seat_a = ['a' + f'{i}' for i in range(1, 31)]
        seat_b = ['b' + f'{i}' for i in range(1, 31)]
        seat_c = ['c' + f'{i}' for i in range(1, 31)]
        seat_d = ['d' + f'{i}' for i in range(1, 31)]
        seat_e = ['e' + f'{i}' for i in range(1, 31)]
        seat_f = ['f' + f'{i}' for i in range(1, 31)]
        seat_g = ['g' + f'{i}' for i in range(1, 31)]
        seat_h = ['h' + f'{i}' for i in range(1, 31)]
        seat_i = ['i' + f'{i}' for i in range(1, 31)]
        seat_j = ['j' + f'{i}' for i in range(1, 31)]
        seat_sum = seat_a + seat_b + seat_c + seat_d + seat_e + seat_f + seat_g + seat_h + seat_i + seat_j

        seat = input('좌석을 선택해 주세요 : ')

        if seat in seat_sum:
            ticket -= 1
            if ticket == 0:
                break
        else:
            print('다시 입력해주세요.')
    except:
        print('다시 입력해주세요.')
                    
print(f'''
        <예매 내역>
         
상영관 : {room_list[room - 1]}
영화   : {movie_list[movie - 1]} / {ticket}매
좌석   : {seat}
시간   : {time_list[time - 1]}
''')
```

```python
# # [과제] 업무미팅이 2시임을 알려주는 자동 이메일 클래스 AutoEmail을 작성하여 아래와 같이 출력하세요.
# # 안녕하세요. Kevin님, 업무미팅은 2시입니다.

class AutoEmail:
    def send(self, name):
        print(f'안녕하세요, {name}님, 업무미팅은 2시입니다.')
        
auto = AutoEmail()
auto.send('kevin')
```
