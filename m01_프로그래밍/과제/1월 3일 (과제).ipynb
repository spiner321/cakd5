```python
import re #정규표현식 패키지 

states = ["Alabama", "Georgia!", 'Georgia', 'georgia', 'FlOrIda', 'south carolina##', 'West virginia?']
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip() # 양쪽 공백제거
        value = re.sub('[!#?]' , "", value) # 특수문자 제거
        value = value.title() # 첫 글자만 대문자로 변경
        result.append(value)
    return result

clean_strings(states)
```

```python
# 연도/성별/출생아수를 구하세요.

def fsx(a):
    sum = 0
    for k in range(len(a)):
        sx = a[k].split(',')[1]
        if sx == "F":
            sum+=int(a[k].split(',')[2])
        else:sum +=0
    return sum

def msx(a):
    sum = 0
    for k in range(len(a)):
        sx = a[k].split(',')[1]
        if sx == "M":
            sum+=int(a[k].split(',')[2])
        else:sum +=0
    return sum

for i in range(1880,2020):
    file = ('D:/python/cakd5/m1_programming/data/names/yob{}.txt').format(i)
    yfile = open(file, 'r')
    a = yfile.read().split('\n')
    del a[-1]
    print(f'{i}년에 출생한 남아는 {msx(a)}명, 여아는 {fsx(a)}입니다.\n')
```

```python
from os.path import exists

def getTop10BabyName(year):
    nameF = {}
    nameM = {}
    
    filename = 'D:/python/cakd5/m1_programming/data/names/yob%s.txt'%year
    if not exists(filename):
        print('[%s] 파일이 존재하지 않습니다.'%filename)
        return None
    
    with open(filename,'r') as f:
        data = f.readlines()
        for d in data:
            if d[-1] == '\n':
                d = d[:-1]
                
                tmp = d.split(',')
                name = tmp[0]
                sex = tmp[1]
                birth = tmp[2]
                
                if sex == 'F':
                    ret = nameF
                else:
                    ret = nameM
                    
                if name in ret:
                    ret[name] += int(birth)
                else:
                    ret[name] = int(birth)
                    
    retF = sorted(nameF.items(), key=lambda x:x[1],reverse=True)
    retM = sorted(nameM.items(), key=lambda x:x[1], reverse=True)

    for i, name in enumerate(retF):
        if i>9:
            break
        print('TOP_%d 여자아기이름:%s'%(i+1,name))
    for i, name in enumerate(retM):
        if i>9:
            break
        print('Top_%d 남자아기이름:%s'%(i+1,name))
y = input('출생년도 입력하세요(예:2001):')
getTop10BabyName(y)
```
