```python
import pandas as pd
import numpy as np
```

```python
df1 = file_data.copy()
df2 = pd.DataFrame({
    '차림표' : ['비빔밥', '돈까스', '라면', '김밥', '제육'],
    '가격' : [7000, 8000, 3000, 3000, 8000]
})
df2
```
