```python
import seaborn as sns

df = sns.load_dataset('titanic')
tdf = df.copy()
tdf.head()
```

```python
# 컬럼이름 변경

tdf.rename(columns = {'sex' : 'gender', 'fare' : 'ticket'}, inplace = True)
tdf.head()
```

```python
# 요소의 개수와 모든 값의 종류를 출력

print(tdf.gender.value_counts()) # 개수
print(tdf.gender.unique()) # 값
```

```python
# 값 변경

df1 = tdf[['gender']].replace(['female', 'male'], [1, 0])
df1.gender.value_counts()
```

```python
# 칼럼 대문자 변경

tdf.rename(str.upper, axis = 'columns', inplace = True)
tdf.head()
```

```python
# 칼럼 첫 글자만 대문자로 변경

tdf.rename(str.title, axis = 'columns', inplace = True)
tdf.tail()
```

```python
# 칼럼 소문자 변경

tdf.rename(str.lower, axis = 'columns', inplace = True)
tdf.head()
```

```python
# 칼럼 순서 변경

columns_customed = ['pclass', 'gender', 'age', 'survived']
tdf[columns_customed].head()
```

```python
# age 칼럼의 범주화

def get_category(age):
    cat = ''
    if age <= 5: cat = 'baby'
    elif age <= 12: cat = 'Child'
    elif age <= 19: cat = 'Teenage'
    elif age <= 35: cat = 'Youngadult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    return cat

tdf['age_cat'] = tdf.age.apply(lambda x: get_category(x))
tdf[['age', 'age_cat']]
```

```python
df_org = pd.DataFrame(boston.data, columns = [boston.feature_names])
df_org['target'] = boston.target
df_org
```

```python
df = df_org[['INDUS', 'RM', 'PTRATIO', 'LSTAT', 'target']]
df
```

```python
# 스케일링

from sklearn.preprocessing import MinMaxScaler # 0 ~ 1 사이의 숫자로 표현하는 비례식

# scaler = MinMaxScaler()
df['RM'] = MinMaxScaler().fit_transform(df.RM)
df.head()
```

```python
li = ['INDUS', 'RM', 'PTRATIO', 'LSTAT', 'target']
for i in li:
    scaler = MinMaxScaler()
    df[i] = scaler.fit_transform(df[[i]])
df.head()
```
