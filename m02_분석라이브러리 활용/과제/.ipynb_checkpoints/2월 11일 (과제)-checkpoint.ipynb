```python
# 쇼핑

import os
import sys
import urllib.request
from bs4 import BeautifulSoup

client_id = "nm2CQVL08k7ff2IEaWua"
client_secret = "ORLP6jy3B6"
input_ = input('검색 : ' )
query = urllib.parse.quote(input_)
display = 100
start = 1
end = 1000

shopping = []
for i in range(start, end, display):
    url = "https://openapi.naver.com/v1/search/shop.xml?query=" + query \
    + '&display=' + str(display) + '&start=' + str(i)
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read().decode('utf-8')
        shopping.append(response_body)
    else:
        print("Error Code:" + rescode)
        
for i in shopping:
    soup = BeautifulSoup(i, 'xml')
    print(soup.prettify())
```

```python
import re

titles = []
for i in shopping:
    soup = BeautifulSoup(i, 'xml')
    title = soup.find_all('title')
    for j in title:
#         titles.append(re.findall('[A-Z0-9]+-[A-Z0-9]+', j.text))
        a = re.sub('<.*?>', '', j.text)
        b = re.sub(f"Naver Open API - shop ::'{input_}'", '', a)
        if len(b) > 0:
            titles.append(b)
        
links = []
for i in shopping:
    soup = BeautifulSoup(i, 'xml')
    link = soup.find_all('link')
    for j in link:
        a = re.sub('https://search.naver.com', '', j.text)
        if len(a) > 0:
            links.append(a)
            
lprices = []
for i in shopping:
    soup = BeautifulSoup(i, 'xml')
    lprice = soup.find_all('lprice')
    for j in lprice:
        lprices.append(j.text)

web_df = pd.DataFrame(columns = ('제품', '최저가(원)', '링크'))

for i in np.arange(len(titles)):
    web_df.loc[i] = [titles[i], lprices[i], links[i]]

web_df
```

```python
city = ['Seoul', 'Tokyo', 'New York']
key = '832fd42874963c9f8cd4afc0283114d6'

for i in city:
    url = f'http://api.openweathermap.org/data/2.5/weather?q={i}&units=metric&APPID={key}'
    data = requests.get(url).json()
    print('---------------------')
    print('+ 도시 =', data['name'])
    print('| 날씨 =', data['weather'][0]['description'])
    print("| 최저 기온 =", data["main"]["temp_min"])
    print("| 최고 기온 =", data["main"]["temp_max"])
    print("| 습도 =", data["main"]["humidity"])
    print("| 기압 =", data["main"]["pressure"])
    print("| 풍향 =", data["wind"]["deg"])
    print("| 풍속 =", data["wind"]["speed"])
    print('---------------------')
    print("")
```

```python
# https://www.data.go.kr/data/15057467/openapi.do - api 신청
import json
import requests
import pandas as pd
import numpy as np
ServiceKey = 'dDUc9SVR/pZIxs5qbu7ABEG4/S5t0ZJBTLZdlveVsR4Hyc2ynwn17cF68xglOrHN/mNu/W4JmkPP3N/M8HF12g==' 
#인증키 - (본인 URL Encode 사용)
searchYear = [2017,2018,2019,2020]
#연도코드 - ( 가능한 년도 옵션이 2017 - 2020 까지만 존재)
siDo = [11,26,27,28,29,30]
#시도코드 - ex(서울 - 11, 부산 - 26, 대구 - 27, 인천 - 28, 광주 - 29, 대전 - 30)
# siDo[0] - 서울
guGun = [680,740,305,500,440,410]
# 시군구코드 ( 시도코드와 시군구코드는 서로 연관이 있어야함 서울시도코드에서 부산구군코드 사용불가능 검색 결과 없음)
# 서울 {강남구 680, 강동구 740, 강북구 305, 강서구 500}
numOfRows = 100 
# 검색건수
pageNo = 1
# 페이지 번호 ( 전체 검색 검수를 확인해 보며 페이지 번호 선택)

car_df = pd.DataFrame(columns=('다발지역지점','사고발생건수','사상자수','경도','위도'))

url = 'http://apis.data.go.kr/B552061/frequentzoneLg/getRestFrequentzoneLg'
params ={'serviceKey' : ServiceKey, 'searchYearCd' : searchYear[0], 'siDo' : siDo[0], 'guGun' : guGun[0], 'type' : 'json', 'numOfRows' : numOfRows, 'pageNo' : pageNo }
response = requests.get(url, params=params)
dictionary = response.json()
count = dictionary['totalCount']
for i in np.arange(count):
    temp = dictionary['items']['item'][i]
    spot_nm = temp['spot_nm']
    occrrnc_cnt = temp['occrrnc_cnt']
    caslt_cnt = temp['caslt_cnt']
    lo_crd = temp['lo_crd']
    la_crd = temp['la_crd']
    car_df.loc[i] = [spot_nm,occrrnc_cnt,caslt_cnt,lo_crd,la_crd]
    
car_df
```
