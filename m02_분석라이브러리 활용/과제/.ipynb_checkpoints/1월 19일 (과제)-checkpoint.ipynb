```python
import math

fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

#=======================히스토그램=======================
num=np.zeros(600)
for i in range(100): num[i]=np.random.randint(1,45)
for i in range(100,250): num[i]=np.random.randint(45,75)
for i in range(250,600): num[i]=np.random.randint(75,100)    
ax1.hist(num, bins = 50, color ='k',alpha = 0.7) 
#=======================선=======================
x=np.arange(360)
y=[]
for i in x:
    a=math.cos(math.pi*(i/180))
    y.append(a)    
ax2.plot(x,y,  color = 'k')
#=======================막대=======================
ax3.bar(x,y,width=0.8)
#=======================산점도=======================
sc_x=np.arange(30)
sc_y=np.full(30,51)
ax4.scatter(sc_x,sc_y)
```

##### [과제] df_auto의 각 칼럼을 시각화해서 탐색한 후 인사이트를 기술하세요(5개 이상)

```python
df = pd.read_excel('auto-mpg.xlsx')
df.loc[(df['horsepower'] == '?'), ['horsepower']] = np.nan
# df = df.convert_dtypes()
df['horsepower'] = df['horsepower'].astype(float)
df.info()
```

```python
# 결측치 채우기

df1 = df.pivot_table(values = 'horsepower', 
                     index = ['displacement', 'weight'],
                     aggfunc = 'mean')

display(df1)

def fill_horsepower(x):
    if np.isnan(x['horsepower']):
        return df1.loc[x['displacement'], x['weight']]
    else:
        return x['horsepower']
    
df.horsepower.fillna(fill_horsepower, inplace = True)
```

```python
plt.figure(figsize = (15, 13))
sns.heatmap(df.corr(), annot_kws = {'size' : 15}, annot = True) 
plt.xticks(size = 12)
plt.yticks(size = 12)
```

```python
# 무게와 배기량은 양의 상관관계

plt.figure(figsize = (8, 8))
plt.scatter(df.displacement, df.weight)
plt.xlabel('displacement')
plt.ylabel('weight')
```

```python
# 배기량이 455인 buick estate wagon (sw)만 무게가 과도하게 가벼움

df[['displacement', 'weight']].describe()
df.loc[(df['weight'] <= 3500) & (df['displacement'] >= 400), 'car name']
df.loc[(df['car name'] == 'buick estate wagon (sw)')]
```

```python
# 대부분의 차가 75~100 마력 사이에 몰려있다.

df.horsepower.plot(kind = 'hist', color = 'red')
```

```python
# 차의 성능 중 가속도만 연비와 상관관계가 보이지 않는다

df = df.sort_values(by = ['mpg'], ascending = False)
df = df.reset_index(drop = True)

fig = plt.figure(figsize = (18, 15))

fig.add_subplot(2, 2, 1)
plt.scatter(df.horsepower, df.mpg, color = 'r')
plt.xlabel('horsepower', size = 15)

fig.add_subplot(2, 2, 2)
plt.scatter(df.displacement, df.mpg, color = 'b')
plt.xlabel('displacement', size = 15)

fig.add_subplot(2, 2, 3)
plt.scatter(df.weight, df.mpg, color = 'k')
plt.xlabel('weight', size = 15)

fig.add_subplot(2, 2, 4)
plt.scatter(df.acceleration, df.mpg)
plt.xlabel('acceleration', size = 15)

# plt.plot(x, y)
```
