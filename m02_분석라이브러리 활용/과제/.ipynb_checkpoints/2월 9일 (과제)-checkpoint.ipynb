##### [과제] 네이버 홈페이지에 있는 ['메일', '카페', '블로그', '지식IN' '쇼핑']을 출력하세요
---

```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://naver.com'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

re.search('메일', html)
print(html[25800:26200], '\n')
a = soup.select('div.group_nav')
b = soup.find_all('a', class_ = ['nav'])[0:5]

c = [i.text for i in b]
print(c)
```

##### [과제] 네이버 영화 랭킹 가져와서 전체 영화제목을 출력하세요

---
- url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnld=105'

```python
import numpy as np
import pandas as pd

url = 'http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnld=105'
html = requests.get(url).text

index = 0
for i in range(len(re.findall('강원도', html))):
    index += re.search('강원도', html[index:]).end()
    print(index)

soup = BeautifulSoup(html[17900:26300], 'html.parser')
gangwon = soup.select('location > province, city, tmEf, tmn, tmx, rnSt')

for i in gangwon:
    print(i)
```

```python
province = [i.text for i in soup.find_all('province')]
city = [i.text for i in soup.find_all('city')]
time = [i.text for i in soup.find_all('tmef')]
# for i in time:
#     i.replace('00:00', '오전') for i in time
#     i.replace('12:00', '오후') for i in time
#     if time[-3:]:
#         i.replace('오전', '')
    
# time = [i.replace('00:00', '오전') for i in time]
# time = [i.replace('12:00', '오후') for i in time ]
tmn = [i.text for i in soup.find_all('tmn')]
tmx = [i.text for i in soup.find_all('tmx')]
rnSt = [i.text for i in soup.find_all('rnst')]

df = pd.DataFrame()
df['지역'] = np.arange(39)
df['도시'] = np.arange(39)
df['일시'] = time
df['최저온도'] = tmn
df['최고온도'] = tmx
df['강수량(%)'] = rnSt

# df.loc[df['일시'] == '2022-02-17 오전', '일시'] = '2022-02-17'
# df.loc[df['일시'] == '2022-02-18 오전', '일시'] = '2022-02-18'
# df.loc[df['일시'] == '2022-02-19 오전', '일시'] = '2022-02-19'

df['지역'].iloc[0:26] = '강원도영서'
df['지역'].iloc[26:] = '강원도영동'
df['도시'].iloc[0:13] = '춘천'
df['도시'].iloc[13:26] = '원주'
df['도시'].iloc[26:] = '강릉'

df1 = df.set_index([df['지역'], df['도시'], df['일시']])
df1.drop(['지역', '도시', '일시'], axis = 1)
```
