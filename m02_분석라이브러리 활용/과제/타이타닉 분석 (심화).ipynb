```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('titanic3.csv')
display(df)
print(df.info())
```

```python
# home.dest 분석

print(len(df['home.dest'].unique()))
df['home.dest'] = df['home.dest'].astype(str)
df[['home.dest']]
plt.figure(figsize = (40, 8))
plt.scatter(x = df['home.dest'], y = df['survived'])
```

```python
# ticket 분석

plt.scatter(y = df['fare'], x = df['ticket'])
```

```python
# 이름 수정

position = df['name'].to_list()
position2 = []
for list_ in position:
    a = list_.split(' ')
    position2.append(a[1][:-1])    
df.name = position2

df.drop(['body', 'cabin', 'boat', 'home.dest', 'ticket'], axis = 1, inplace = True)
df.rename(columns = {'name' : 'position'}, inplace = True)
```

```python
df.info()
df.head()
```

```python
# embarked 숫자로 변경

print(df['embarked'].unique())
df['embarked'] = df['embarked'].replace(['S', 'C', 'Q'], [0, 1, 2])
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
# 나이 결측치 채우기

df_age = df.pivot_table(values = 'age', index = ['pclass', 'survived'], aggfunc = 'mean')
def fill_age(x):   
    if np.isnan(x['age']):
        return df_age.loc[x['pclass'], x['survived']]
    else:
        return x['age']
df['age'] = df.apply(fill_age, axis = 1).astype(float).round(2)
```

```python
# 요금 결측치 채우기

df_fare = df.pivot_table(values = 'fare', index = ['pclass'], aggfunc = 'mean')
def fill_fare(x):
    if np.isnan(x['fare']):
        return df_fare.loc[x['pclass']]
    else:
        return x['fare']
df['fare'] = df.apply(fill_fare, axis = 1).astype(float)
```

```python
# 호칭 통일

print(df['position'].unique())
df1 = df.pivot_table(values = ['age', 'survived'],
                     index = ['position', 'sex'], 
                     aggfunc = 'mean')
display(df1)

df2 = df.loc[(df['position'] == 'Ms') | (df['position'] == 'Miss') | (df['position'] == 'Mrs')]
df3 = df2.pivot_table(values = ['age'],
                     index = ['position'], 
                     aggfunc = ['min', 'max', 'mean'])
df3
```

```python
df4 = df.loc[(df['position'] != 'Ms') & (df['position'] != 'Miss') & (df['position'] != 'Mrs') & (df['position'] != 'Mr')]
print(df4['position'].value_counts(), '\n')

df5 = df.loc[(df['position'] != 'Mr') & (df['sex'] == 'male') & (df['survived'] == 1)]
print(df5['position'].value_counts())
```

```python
plt.figure(figsize = (25, 8))

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

plt.bar(df4['position'].value_counts().index, list(df4['position'].value_counts()), width = 0.8, label = '모든 기타 호칭') # 모든 기타 호칭
plt.bar(df5['position'].value_counts().index, list(df5['position'].value_counts()), width = 0.8, label = '기타 호칭 중에서 생존한 남성') # 기타 호칭 중에서 생존한 남성
plt.xticks(rotation = 45, size = 15)
plt.legend(fontsize = 15)
```

```python
df.loc[(df['sex'] == 'male') & (df['position'] != 'Master'), 'position'] = 'Mr'
df.loc[(df['sex'] == 'female') & (df['sibsp'] > 0) & (df['position'] != 'Mrs') & (df['position'] != 'Miss'), 'position'] = 'Mrs'
df.loc[(df['sex'] == 'female') & (df['sibsp'] == 0) & (df['position'] != 'Mrs') & (df['position'] != 'Miss'), 'position'] = 'Miss'
df.position.value_counts()
```

```python
# embarked 결측치 채우기

df_embarked = df.loc[(df['sex'] == 'female') & (df['survived'] == 1)].pivot_table(values = ['age'], index = ['embarked', 'sex', 'position'], aggfunc = ['mean', 'max'])
display(df_embarked)
df.loc[df['embarked'].isnull()]
```

```python
df.loc[df['embarked'].isnull(), 'embarked'] = 1
df.loc[df['embarked'].isnull()]
```

```python
# 모든 데이터 숫자화

from sklearn.preprocessing import LabelEncoder

features = ['sex', 'position']

for feature in features:
    df[feature] = LabelEncoder().fit_transform(df[feature])
    
df
```

```python
# 나이와 생존률

df_age = df[['age', 'survived']].describe(percentiles = np.arange(10) * 0.1)
plt.figure(figsize = (10, 7))
plt.plot(df_age.index[4:], df_age.survived[4:])
plt.xlabel('age', size = 20)
df_age
```

```python
# 승객등급과 생존률

df_pclass = pd.pivot_table(df,
                           values = 'survived',
                           index = 'pclass',
                           aggfunc = 'mean')

display(df_pclass)

ratio = []
for i in range(len(df_pclass['survived'])):
    a = df_pclass['survived'].iloc[i] / df_pclass['survived'].sum()
    ratio.append(a)
labels = df_pclass.index

plt.figure(figsize = (10, 10))
plt.pie(ratio, autopct = '%.1f%%', textprops = {'fontsize' : 15}, startangle = 0)
plt.legend(labels = df_pclass.index, fontsize = 20)
```

```python
from sklearn.model_selection import train_test_split

y_df = df['survived'] # 종속 변수
x_df = df.drop('survived', axis = 1) # 독립 변수

x_train, x_test, y_train, y_test = train_test_split(x_df, y_df, test_size = 0.2, random_state = 7)

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
accuracy_rf = accuracy_score(y_test, rf_pred).round(2)

lr = LogisticRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
accuracy_lr = accuracy_score(y_test, lr_pred).round(2)

print(f'rf 정확도 : {accuracy_rf}, lr 정확도 : {accuracy_lr}')
```
