##### [과제] 인덱싱을 이용하여 특정 컬럼, 특정 행, 특정 값을 추출

```python
data = np.random.choice(range(1, 26), (5, 5), replace = False)
# data = np.arange(1, 26).reshape((5, 5))
columns = [f'가로{i}' for i in range(1, 6)]
index = [f'세로{i}' for i in range(1, 6)]
df = pd.DataFrame(data, index, columns)
df
```

```python
print(df['가로1'], '\n')
print(df[['가로1', '가로4']], '\n')
print(df.iloc[1, 3], '\n')
print(df.iloc[[1, 3]], '\n')
```

```python
columns = ['kor','eng','art','math','gym','sci']
index = ['jim','sam','tom','tim','bob']
data = np.random.randint(70,100,size=(5,6))
df = pd.DataFrame(data,index,columns=columns)
df
```

```python
print(df['kor'], '\n')
print(df.loc['sam', 'math'])
print(df.iloc[1, 3], '\n')
print(df.loc['tom', ['eng', 'sci']])
print(df.iloc[2, [1, 5]], '\n')
print(df.loc[['jim', 'tim'], ['eng', 'gym']], '\n')
print(df.iloc[0:2, 3:6])
```

##### [과제] 데이터 추출(인덱싱, 조건), 변경, 삭제, 정렬, 칼럼 병합(sibsp + parch = family) 등을 수행하여 아래의 데이터에서 인사이트를 찾아보세요 
https://blog.naver.com/sjy5448/222391446184

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df1 = df.copy().head()
df1.drop(['survived', 'pclass', 'embarked', 'deck', 'parch', 'alone', 'who', 'adult_male'], axis = 1, inplace = True)
df1['family'] = df['sibsp'] + df['parch']
df1
```

```python
print(df1.loc[0:, ['sex', 'alive']].sort_values(by = 'alive', ascending = False), '\n') # 여성만 생존
print(df1.loc[0:, ['fare', 'class']].sort_values(by = 'fare', ascending = False), '\n') # Firstclass 와 Thirdclass의 가격은 자릿수가 다르다.

df_sibsp = df1.loc[df1['class'] == 'First']
df_sibsp = df1.loc[df1['sex'] == 'female']
print(df_sibsp.drop(['fare', 'embark_town', 'alive', 'family'], axis = 1).sort_values(by = 'age'), '\n') # 30세 이상의 배우자가 있는 여성만 Firstclass 탑승
```
