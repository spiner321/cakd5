##### [과제] 10을 더하는 함수 add_10 과 두 객체의 합 add_two_obj 함수를 정의한 후 결과값이 각각 20과 20을 출력하세요.

```python
class Cal:
    def add_10(self, x):
        return x + 10
    def add_two_obj(self, x, y):
        return x + y
    
cal = Cal()
print(cal.add_10(10))
print(cal.add_two_obj(15, 5))
```

##### [과제] 10을 곱하는 사용자 함수와 두 객체의 나눗셈을 수행하는 사용자 함수를 정의한 후 결과값이 각각 200과 10을 출력하세요.

```python
def mul(x):
    return x * 10
def div(x, y):
    return x / y

print(mul(20))
print(div(100, 10))
```
