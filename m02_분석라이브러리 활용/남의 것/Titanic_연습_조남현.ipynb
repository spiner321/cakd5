```python
import seaborn as sns
import pandas as pd
import numpy as np
df = sns.load_dataset('titanic')
# 데이터를 전체적으로 훑어봄       >>> null값이 있음.
df.head(50)
```

```python
# 데이터들의 정보를 요약해서 보고 의미 있는 column을 확인
# 결측치가 데이터에 영향이 크다면, 전처리를 해야 함.

# SibSp : 타이타닉 호에 탑승한 형제/배우자 수
# Parch : 타이타닉 호에 탑승한 부모/자녀 수
# Embark : 승선항
# Pclass : 티켓 등급
df.info()
```

```python
# 작업할 데이터프레임을 하나 만들기
new_df = df.drop(['fare','embarked','deck','embark_town','alive','alone','adult_male','class'],axis = 1)
new_df
```

```python
# 나이의 null을 채워주기 위해 기준을 정함. pclass가 영향이 클 것으로 예상
meanage_pclass = new_df.pivot_table(values = 'age', index = 'pclass', aggfunc = 'mean') # index별 value의 평균
print(meanage_pclass)
age_ori = new_df[new_df.age.isnull()]                                                   # age의 null값 가진 row들을 추출
new_df_pc1 = age_ori[age_ori.pclass == 1] ; new_df_pc1 = new_df_pc1.fillna(value=38)    # 각 pclass마다 평균 나이로 대체
new_df_pc2 = age_ori[age_ori.pclass == 2] ; new_df_pc2 = new_df_pc2.fillna(value=30)
new_df_pc3 = age_ori[age_ori.pclass == 3] ; new_df_pc3 = new_df_pc3.fillna(value=25)
new_df_drop = new_df.dropna(subset = ['age'])                                           # null이 있는 age 열 삭제
new_df_concat = pd.concat([new_df_drop,new_df_pc1,new_df_pc2,new_df_pc3])               # 대체한 열로 재조합
df = new_df_concat                                                                      # 확인 후 데이터 프레임을 바꿈
df.info()
```

```python
df.groupby(['pclass','sex']).mean()     # pclass와 sex를 함께 기준으로 잡고 평균을 봄
```

## 정리
1. pclass는 생존할 확률에 영향을 끼침
    - 높은 등급일 수록 생존 확률 높음
2. 여자가 남자보다 생존할 확률이 높음
    - 너무 큰 차이이므로 전체 평균에서도 마찬가지인 듯 함
3. 평균 나이에서 여자는 더 어릴수록, 남자는 더 나이가 많을 수록 생존 확률이 높음
    - 왜 다른지 검증 필요..
4. 여자는 같은 위치에 있는 동반자가 있을 수록, 남자는 없을 수록 생존율이 높아보임
    - 검증 필요
5. 같은 위치의 동반자가 있을 때보다 챙겨야 할 동반자가 있을 때 생존율이 떨어짐

## 결론 (가설)
- 현재까지 분석한 결과 생존율이 높기 위한 조건은 다음과 같음
    - 높은 등급의 좌석, 성별은 여자
    - 남자의 경우: 나이가 많을수록, 동반자가 없거나, 있어도 같은 위치의 사람일 수록
    - 여자의 경우: 나이가 적을수록, 동반자가 있고, 같은 위치의 사람일 수록  
→ 높은 등급의 좌석이 위험 상황에 대비해 위치적 요소나 안전장비, 서비스 등의 요소가 좋아서 생존율이 높은 듯 함  
→ 동반자가 있는 경우, 먼저 챙겨야 할 인원이 있는 경우 사망율이 높아지는 듯 함  
→ 특히, 남성은 동반자가 없을 수록 생존율이 현저히 높음  
→ 위기 상황에 남성이 대비를 못해 죽거나, 위험을 무릎쓰고 돕기 때문에 사망률이 높은 듯 함
