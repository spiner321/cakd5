```python
import numpy as np
import pandas as pd
from datetime import datetime
from random import randrange
import warnings
warnings.filterwarnings(action='ignore')
index=np.arange(10000)
df=pd.DataFrame(np.random.rand(10000,11),index=index,columns=['cusno','gender','age','region',
                                           'day','date','product1','price','quantity','time','amount']) 

df.cusno=np.random.randint(0,1000,10000)
gender=np.random.randint(0,2,10000)
for i in range(len(df)):
    df.gender[i]=gender[df.cusno[i]]
age=np.random.randint(0,81,10000) 
for i in range(len(df)):
    df.age[i]=age[df.cusno[i]]
region=np.random.randint(0,11,10000)
for i in range(len(df)):
    df.region[i]=region[df.cusno[i]]
date_list=[]
start_date = pd.to_datetime('2020-01-01')
end_date = pd.to_datetime('2021-12-31')
date_list = pd.date_range(start_date,end_date,freq='D')
df.date=np.random.choice(date_list,10000)
df.day = df.date.dt.day_name()
df.product1=np.random.randint(10,40,10000)
price= np.random.randint(1,50,10000)
for i in range(len(df)):
    if df.product1[i] < 20:
        df.price[i] = price[df.product1[i]]*10
    elif df.product1[i] < 30:
        df.price[i] = price[df.product1[i]]*100
    elif df.product1[i] < 40:
        df.price[i] = price[df.product1[i]]*1000
df.quantity=np.random.randint(1,50,10000)
# df.quantity_21=np.random.randint(1,101,10000)
df.time=np.random.randint(10,23,10000)
# df.time_21=np.random.randint(1,25,10000)
df.amount = df.price*df.quantity
# df.amount_21 = df.price_21*df.quantity_21
# df.loc[df.amount_21 > df.amount_20,'sales'] = 1
# df.loc[df.amount_21 < df.amount_20,'sales'] = 0
# df.loc[df.amount_21 == df.amount_20,'sales'] = np.nan
df
```

```python
# 시간 10 to 22 입장시간 
# 주말 매출 증가
# 제품별 가격대 설정 저 중 고 10  20  30
# 크리스마스등 특정 연휴에 판매 증가 저->고 갈수록 제품의 매출 증가 
# 월-금 시간대에 시간을 제약해보기 (20대 ~ 50대 )
# 나이대에 따라 저 중 고 선호도 추가
df1=df.copy()
df1.time_20=np.random.randint(10,23,10000)
df1.time_21=np.random.randint(10,23,10000)
df1['year']=df1.date.dt.year
for i in range(len(df1)):
    if df1.age[i] < 10:
        df1.age[i] = 0
    elif df1.age[i] < 20:
        df1.age[i] = 10
    elif df1.age[i] < 30:
        df1.age[i] = 20
    elif df1.age[i] < 40:
        df1.age[i] = 30
    elif df1.age[i] < 50:
        df1.age[i] = 40
    elif df1.age[i] < 60:
        df1.age[i] = 50
    elif df1.age[i] < 70:
        df1.age[i] = 60
    elif df1.age[i] < 80:
        df1.age[i] = 70
df1['who']=df1.age
for i in range(len(df1)):    
    if df1.who[i] < 10:
        df1.who[i] = 'child'
    elif df1.who[i] < 25:
        df1.who[i] = 'student' 
    elif df1.who[i] < 65:
        df1.who[i] = 'worker'
    else: 
        df1.who[i] = 'elder'
# df1.describe()
df1
```
