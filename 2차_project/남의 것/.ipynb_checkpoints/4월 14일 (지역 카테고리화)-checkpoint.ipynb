```python
import pymysql
import pandas as pd
import re
```

```python
connection = pymysql.connect(host='127.0.0.1', user='root', password='root', db='smart_factory', charset='utf8')   # 접속정보
cursor = connection.cursor()  
sql = "SELECT 제품코드, 제품명 FROM `product`"   
cursor.execute(sql)  
result=cursor.fetchall()
columns = ['제품코드', '제품명']
df = pd.DataFrame(result, columns = columns)
connection.commit()
connection.close()
```

# # product 테이블에서 제품명 컬럼을 제품과 지역으로 분리
---

```python
products = []
for i in df['제품명']:
    a = re.search('[a-zA-Z]+-*[0-9a-zA-Z]+', i).group()
    products.append(a)
    
products
```

```python
regions = []
for i in df['제품명']:
    a = re.search(r'\([가-힣]*\)', i)
    if a == None:
        a = re.search('[가-힣]+', i)
        regions.append(a.group())
        
    else:
        b = re.sub(r'\(*\)*', '', a.group())
        regions.append(b)

regions
```

```python
df['제품'] = products
df['지역'] = regions
```

```python
def cat_regions(x):
    cat = ''
    if x == '서울' or x == '강서' or x == '송파' or x == '남서울' or x == '동서울' or x == '풍납' or x == '성수': cat = '서울'
    elif x == '동두천' or x == '남양주' or x == '파주' or x == '덕소' or x == '양주' or x == '일산' or x == '의정부' or x == '김포' or x == '진접' or x == '포천' or x == '구리' or x == '고양' or x == '화도': cat = '경기도북부'
    elif x == '하남' or x == '용인' or x == '광주' or x == '오산' or x == '수원' or x == '평택' or x == '이천' or x == '동탄' or x == '안성' or x == '양평' or x == '화성' or x == '여주' or x == '안산' or x == '광명' or x == '성남' or x == '비봉' or x == '의왕' or x == '미사리' or x == '안양': cat = '경기도남부'
    elif x == '인천' or x =='가좌' or x =='부천' or x =='송도' or x =='서인천' or x =='대산': cat = '인천'
    elif x == '부산' or x =='정관' or x =='회동동' or x =='서부산': cat = '부산광역시'
    elif x == '울산' or x =='언양' or x =='온산' or x =='남부': cat = '울산광역시'        
    elif x == '대구' or x =='하빈': cat = '대구광역시'
    elif x == '포항' or x =='경주' or x =='칠곡' or x =='울진' or x =='문경' or x =='안동' or x =='약목' or x =='경산' or x =='단촌' or x =='고령' or x =='영덕' or x =='예천' or x =='구미' or x =='영양' or x =='성주' or x =='남포항': cat = '경상북도' 
    elif x == '함안' or x =='진해' or x =='김해' or x =='의령' or x =='양산' or x =='합천' or x =='칠서' or x =='진주' or x =='산청' or x =='함양' or x =='창녕' or x =='창원' or x =='밀양' or x =='하동' or x =='진영' or x =='사천' or x =='마산': cat = '경상남도' 
    elif x == '정읍' or x =='이서' or x =='익산' or x =='완주' or x =='군산' or x =='남원' or x =='부안' or x =='장수' or x =='임실' or x =='전주' or x =='오수' or x =='군위' or x =='순창': cat = '전라북도'
    elif x == '순천' or x =='담양' or x =='여수' or x =='나주' or x =='영광' or x =='화순' or x =='장흥' or x =='해남' or x =='광양' or x =='노화도' or x =='목포' or x =='장성' or x =='무안' or x =='구례' or x =='군산한전주': cat = '전라남도'
    elif x == '대전' or x =='장동': cat = '대전광역시'
    elif x == '세종' or x =='연기': cat = '세종시'
    elif x == '음성' or x == '청주' or  x == '청원' or x =='단양' or x == '옥산' or x == '영동' or x == '청주금성' or x == '옥계' or x == '보은' or x == '진천' or x == '청양' or x == '충주' or x == '옥천': cat = '충청북도'
    elif x == '홍성' or x == '서산' or x == '천안' or x == '대산' or x == '당진' or x == '금산' or x == '논산' or x == '태안' or x == '부여' or x == '예산' or x == '천북' or x == '보령' or x == '계룡' or x == '공주' or x == '아산': cat = '충청남도'
    elif x == '신림' or x =='철원' or x =='양양' or x =='영월' or x =='쌍용' or x =='평창' or x == '고성' or x == '횡성' or x == '삼척' or x == '강릉' or x == '옥계' or x == '동해' or x == '춘천' or  x =='원주': cat = '강원도'
    else : cat = 0
    
    return cat

df1 = df.copy()    
df1['cat_regions'] = df.지역.apply(cat_regions)

df1.head()
```
