# # 실데이터 가져오기

```python
# 실제 데이터를 가져오는 사용자함수
def get_data():
    import pandas as pd
    import mariadb
    import sys

    # Connect to MariaDB Platform
    try:
        conn = mariadb.connect(
            user="root",
            password="root",
            host="localhost",
            port=3306,
            database="smart_factory"
        )
    except mariadb.Error as e:
        print(f"Error connecting to MariaDB Platform: {e}")
        sys.exit(1)
    
    # Get Cursor
    cur = conn.cursor()
    cur.execute(f"select 수주일자, 거래처코드, 제품명, `contract`.제품코드, 지역16, 지역, 수주량 from `contract`, `production_re` where `production_re`.`제품코드` = `contract`.`제품코드`")
    x = cur.fetchall()
    df = pd.DataFrame(x,columns=['SOLDDATE','CUSTID','PRODNAME','PRODCODE','REGION_U','REGION','QUANT'])
    df.loc[df.REGION_U == '0','REGION'] = '태안'
    df.loc[df.REGION_U == '0','REGION_U'] = '충청남도'
    df.QUANT = df.QUANT.astype(int)
    df = df.sort_values('SOLDDATE').reset_index(drop=True)
    return df
```

# # 빈 날짜들 만들기

```python
def get_date(startdate = None, enddate=None):
    import datetime
    import numpy as np
    import pandas as pd 
    date = startdate

    li = []
    stop = 0
    stop_year = date.year + 1
    while stop == 0:
        li.append(date)
        date = date + datetime.timedelta(days=1)
        if date.year == stop_year:
            stop = 1
        elif date - datetime.timedelta(days=1) == enddate:
            stop = 1
    df = pd.DataFrame(np.array(li),columns=['SOLDDATE'])
    # df.SOLDDATE = df.SOLDDATE.apply(lambda x: str(x.date()))
    return df

def get_random_date(df = None):
    import datetime
    import pandas as pd
    import numpy as np

    df.SOLDDATE = df.SOLDDATE.astype('datetime64')
    year = df.SOLDDATE.min().year
    df_merged = df[['SOLDDATE','CUSTID','PRODNAME','REGION','QUANT']]
    maxdate = df.SOLDDATE.max()
    mindate = df.SOLDDATE.min()
    between = (df.SOLDDATE.max() - df.SOLDDATE.min()).days

    # 뒷부분 채우기
    startdate = maxdate ; enddate = maxdate
    while 1:
        len_date = len(df.SOLDDATE) + np.random.randint(len(df)*-0.1,len(df)*0.1) # 52개 날짜들에 대한 데이터 수에 오차를 줌
        enddate += datetime.timedelta(between + 1)
        startdate = enddate - datetime.timedelta(between)

        # 시작일이 전년으로 가면 해당 연의 첫날로 배정, 데이터 갯수도 비율 맞춰줌
        if enddate.year > year:
            enddate = datetime.datetime(year,12,31)
            len_date = round(len_date * (enddate - startdate).days / between)

        df_temp = get_date(startdate = startdate, enddate = enddate)
        temp_between = (enddate - startdate).days

        # 해당 기간 중 빠지는 날짜를 고려하고, 선택된 날짜 중에서 갯수만큼 선택
        temp_array = np.random.choice(df_temp.SOLDDATE.unique(),size = temp_between - round(np.random.randint(6,9)*temp_between / between),replace = False)
        df_temp2 = pd.DataFrame(np.random.choice(temp_array,size = len_date),columns=['SOLDDATE'])

        # 제품명은 비율을 유지하여 뽑아주도록 함
        if len_date <= len(df.PRODNAME):
            df_temp2['PRODNAME'] = np.random.choice(df.PRODNAME,size = len_date,replace=False)
        else:
            df_temp2['PRODNAME'] = np.random.choice(df.PRODNAME,size = len_date,replace=True)
        df_temp2 = df_temp2.sort_values('SOLDDATE').reset_index(drop=True)
        df_merged = pd.concat([df_merged,df_temp2],axis=0)
        if df_merged.SOLDDATE.max().date() == datetime.datetime(year,12,31).date(): break

    # 앞부분 채우기
    startdate = mindate ; enddate = mindate
    while 1:
        len_date = len(df.SOLDDATE) + np.random.randint(len(df)*-0.05, len(df)*0.05) # 52개 날짜들에 대한 데이터 수에 오차를 줌
        startdate -= datetime.timedelta(between + 1)
        enddate = startdate + datetime.timedelta(between)

        # 시작일이 전년으로 가면 해당 연의 첫날로 배정, 데이터 갯수도 비율 맞춰줌
        if startdate.year < year:
            startdate = datetime.datetime(year,1,1)
            len_date = round(len_date * (enddate - startdate).days / between)
            
        df_temp = get_date(startdate = startdate, enddate = enddate)
        temp_between = (enddate - startdate).days

        # 해당 기간 중 빠지는 날짜를 고려하고, 선택된 날짜 중에서 갯수만큼 선택
        temp_array = np.random.choice(df_temp.SOLDDATE.unique(),size = temp_between - round(np.random.randint(6,9)*temp_between / between),replace = False)
        df_temp2 = pd.DataFrame(np.random.choice(temp_array,size = len_date),columns=['SOLDDATE'])
        
        # 제품명은 비율을 유지하여 뽑아주도록 함
        if len_date <= len(df.PRODNAME):
            df_temp2['PRODNAME'] = np.random.choice(df.PRODNAME,size = len_date,replace=False)
        else:
            df_temp2['PRODNAME'] = np.random.choice(df.PRODNAME,size = len_date,replace=True)
        df_temp2 = df_temp2.sort_values('SOLDDATE').reset_index(drop=True)
        df_merged = pd.concat([df_temp2,df_merged],axis=0)
        if df_merged.SOLDDATE.min().date() == datetime.datetime(year,1,1).date(): break

    df.SOLDDATE = df.SOLDDATE.astype('str')
    df_merged.SOLDDATE = df_merged.SOLDDATE.astype('str')

    # 랜덤 지역 부여, 그에 맞는 대표지역과 코드 불러오기
    df_merged.loc[df_merged.REGION.isnull() == True,'REGION'] = np.random.choice(df.REGION.unique(),size=df_merged.REGION.isnull().sum())
    df_merged = df_merged.merge(df[['REGION','REGION_U']].drop_duplicates(), how = 'inner',on = 'REGION')
    
    # 랜덤 고객번호 생성, 고객번호는 날짜 내에서 곂칠 수 있도록 하였음.
    # 없도록 할 수 있으니 추가 확인이 필요함.
    for i in df_merged.SOLDDATE.unique():
        df_merged.loc[(df_merged.SOLDDATE == i)&(df_merged.CUSTID.isnull() == True),'CUSTID'] = np.random.choice(df.CUSTID.unique(), size = len(df_merged.loc[(df_merged.SOLDDATE == i)&(df_merged.CUSTID.isnull() == True),'CUSTID']),replace=False)
    df_merged = df_merged[['SOLDDATE', 'CUSTID', 'PRODNAME', 'REGION_U', 'REGION']]
    df_merged = df_merged.sort_values(by='SOLDDATE').reset_index(drop=True)

    # 없도록 하려면 위 대신 밑에 코드로 사용.
    # for i in df_merged.SOLDDATE.unique():
    #     df_merged.loc[(df_merged.SOLDDATE == i)&(df_merged.CUSTID.isnull() == True),'CUSTID'] = np.random.choice(df.CUSTID.unique(),size=len(df_merged.loc[(df_merged.SOLDDATE == i)&(df_merged.CUSTID.isnull() == True),'CUSTID'],replace=False))
    # df_merged = df_merged[['SOLDDATE', 'CUSTID', 'PRODNAME', 'REGION_U', 'REGION']]
    # df_merged = df_merged.sort_values(by='SOLDDATE').reset_index(drop=True)

    
    # QUANT 컬럼을 가져온다.
    df_merged= df_merged.merge(df[['SOLDDATE','CUSTID','PRODNAME','REGION','QUANT']], how = 'left',on = ['SOLDDATE','CUSTID','PRODNAME','REGION'])

    return df_merged
```

```python
# ## 수도권
# 서울 = ['서울','강서','송파','남서울','동서울','신림','풍납','성수']  ### 기상CODE 108 / (서울) 수도권기상청
# 경기도북부 = ['동두천','남양주','파주','덕소','양주','일산','의정부','김포','진접','포천','구리','고양','화도']  ### 기상CODE 98 / (동두천) 수도권기상청
# 경기도남부 = ['하남','용인','광주','오산','수원','평택','이천','동탄','안성','양평','화성','여주','안산','광명','성남','비봉','의왕','미사리']   ### 기상CODE 119 / (수원) 수도권기상청
# 인천 = ['인천','가좌','부천','송도','서인천','대산']  ### 기상CODE 112 / (인천) 수도권기상청

# ## 경상도 
# 부산광역시 = ['부산','정관','회동동','서부산']       ### 기상CODE / 159 (부산) 부산지방기상청
# 울산광역시 = ['울산','언양','온산','남부']  ### 기상CODE / 152 (울산) 울산기상대	
# 대구광역시 = ['대구','하빈']                ### 기상CODE / 143 (대구) 대구지방기상청
# 경상북도 = ['포항','경주','칠곡','울진','문경','안동','약목','경산','단촌','고령','영덕','예천','구미','영양','성주','남포항' ]  ### 기상CODE 278 / (의성) 대구지방기상청
# 경상남도 =['함안','진해','김해','의령','양산','합천','칠서','진주','산청','함양','창녕','창원','밀양','하동','진영','사천','마산']  ### 기상CODE 263 / (의령군) 창원기상대


# ## 전라도
# 전라북도 =['정읍','이서','익산','완주','군산','남원','부안','장수','임실','전주','오수','군위','순창' ]  ### 기상CODE 146 / (전주) 전주기상지청
# 전라남도 =['순천','담양','여수','나주','영광','화순','장흥','해남','광양','노화도','목포','장성','무안','구례','군산한전주']   ### 기상CODE 156 / (광주) 광주지방기상청


# ## 충청도
# 대전광역시 = ['대전','장동']  ### 기상CODE 133 / (대전) 대전지방기상청
# 세종시 = ['세종','연기']  ### 기상CODE 239 / (세종) 대전지방기상청
# 충청북도 = [ '청주','음성','충주','옥천','청원','진천','보은','단양','옥산' ]  ### 기상CODE 131 / (청주) 청주기상지청	
# 충청남도 = [ '서산','당진','계룡','천안','아산','영동','청양','홍성','금산','태안','공주','예산','천북','부여','논산','보령' ]  ### 기상CODE 129 / (서산) 홍성기상대


# ## 강원도
# 강원도 = ['강릉','동해','원주','삼척','철원','횡성','옥계','고성','양양','평창','춘천','영월']  ### 기상CODE 114 / (원주) 강원지방기상청	

def cat_location(x):
        if x == '서울': return 108
        elif x == '경기도북부': return 98
        elif x == '경기도남부': return 119
        elif x == '인천': return 112
        elif x == '부산광역시': return 159
        elif x == '대구광역시': return 143
        elif x == '울산광역시': return 152
        elif x == '경상북도': return 278
        elif x == '경상남도': return 263
        elif x == '전라북도': return 146
        elif x == '전라남도': return 156
        elif x == '대전광역시': return 133
        elif x == '세종시': return 239
        elif x == '충청북도': return 131
        elif x == '충청남도': return 129
        elif x == '강원도': return 114
        else: return 119 ## 해당 안될시 생판지역인 평택 기준 날씨로 변환

def loc_weather(start_date, end_date, location):
    import pandas as pd
    import requests
    import numpy as np

    start_date = start_date.replace('-','')
    end_date = end_date.replace('-','')

    location_code = cat_location(location)

    url = 'http://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList'

    params ={'serviceKey' : 'ZKOx0KH7l+PcSZZNRvuI54pjFf5gbYeIa1ccvoUcbzlwPA7ZRd9AqYB+V6++N/urN+9OncLmDH9MvqvMu5SKbg==', 
            'pageNo' : '1', 
            'numOfRows' : '999', 
            'dataType' : 'JSON', 
            'dataCd' : 'ASOS', 
            'dateCd' : 'DAY', 
            'startDt' : start_date,
            'endDt' : end_date, 
            'stnIds' : str(location_code) }


    response = requests.get(url, params=params).json()
    r_response = response.get("response")
    r_body = r_response.get("body")
    r_items = r_body.get("items")
    r_item = r_items.get("item")

    time = [] #일자
    tem = [] #온도
    hum = [] #습도

    for i in range(len(r_item)):
        time.append(r_item[i]['tm'])
        tem.append(r_item[i]['avgTa'])
        hum.append(r_item[i]['avgRhm'])

    time = pd.Series(time)
    tem = pd.Series(tem)
    hum = pd.Series(hum)

    data = [time, tem, hum]

    df = pd.concat(data, axis=1)

    df.columns = ['SOLDDATE','TEMP','HUM']
    df.replace('', 0,inplace=True)

    df['REGION_U'] = location
    df['SOLDDATE'] = pd.to_datetime(df['SOLDDATE'])
    df['TEMP'] = round(df['TEMP'].astype('float32'),1)
    df['HUM'] = round(df['HUM'].astype('float32'),1)

    return df

def make_weather_data(year):
    start_date = str(year) + '-01-01'
    end_date = str(year) + '-12-31'
    loc_li = ['서울','경기도북부','경기도남부','인천','부산광역시','대구광역시','울산광역시','경상북도','경상남도','전라북도','전라남도','대전광역시','세종시','충청북도','충청남도','강원도']
    globals()[f'weather_data_{year}'] = pd.DataFrame()
    
    for i in loc_li:
        globals()[f'weather_{i}'] = loc_weather(start_date,end_date,i)
        globals()[f'weather_data_{year}'] = pd.concat([globals()[f'weather_data_{year}'], globals()[f'weather_{i}']], axis=0)
    
    globals()[f'weather_data_{year}'].reset_index(drop=True,inplace=True)
    globals()[f'weather_data_{year}'].to_csv(f'./dataset/weather_data_{year}.csv', index=False)
    return globals()[f'weather_data_{year}']
```

```python
# CONSTRUCTION 가져오기
def get_construction():
    import pandas as pd
    import mariadb
    import sys

    # Connect to MariaDB Platform
    try:
        conn = mariadb.connect(
            user="root",
            password="root",
            host="localhost",
            port=3306,
            database="smart_factory"
        )
    except mariadb.Error as e:
        print(f"Error connecting to MariaDB Platform: {e}")
        sys.exit(1)
    
    # Get Cursor
    cur = conn.cursor()
    cur.execute(f"select 연도, 월, 총계, 전월대비증감율 from `building_construction`")
    x = cur.fetchall()
    df = pd.DataFrame(x,columns=['YEAR','MONTH','CONSTRUCTION','CONSTRUCTION_RATE'])
    return df

# PURMISSION 가져오기
def get_permission():
    import pandas as pd
    import mariadb
    import sys

    # Connect to MariaDB Platform
    try:
        conn = mariadb.connect(
            user="root",
            password="root",
            host="localhost",
            port=3306,
            database="smart_factory"
        )
    except mariadb.Error as e:
        print(f"Error connecting to MariaDB Platform: {e}")
        sys.exit(1)
    
    # Get Cursor
    cur = conn.cursor()
    cur.execute(f"select 연도, 월, 총계, 전월대비증감율 from `building_permission`")
    x = cur.fetchall()
    df = pd.DataFrame(x,columns=['YEAR','MONTH','PERMISSION','PERMISSION_RATE'])
    return df

# 병합
def concat_all(df,weather_data):
    import pandas as pd
    
    # weather_data 병합
    weather_data.SOLDDATE = weather_data.SOLDDATE.astype('str')
    df.SOLDDATE = df.SOLDDATE.astype('str')
    df = df.merge(weather_data, how = 'inner', on = ['SOLDDATE', 'REGION_U'])
    
    # construction 병합
    df_construction = get_construction()
    df['CONSTRUCTION'] = 0
    df['CONSTRUCTION_RATE'] = 0
    df.SOLDDATE = df.SOLDDATE.astype('datetime64')
    for i in df.index:
        year = df.loc[i,'SOLDDATE'].year
        month = df.loc[i,'SOLDDATE'].month
        df.loc[i,'CONSTRUCTION'] = df_construction[(df_construction['YEAR']==year) & (df_construction['MONTH']==month)].CONSTRUCTION.values[0]
        df.loc[i,'CONSTRUCTION_RATE'] = df_construction[(df_construction['YEAR']==year) & (df_construction['MONTH']==month)].CONSTRUCTION_RATE.values[0]
    
    df_permission = get_permission()
    df['PERMISSION'] = 0
    df['PERMISSION_RATE'] = 0
    for i in df.index:
        year = df.loc[i,'SOLDDATE'].year
        month = df.loc[i,'SOLDDATE'].month
        df.loc[i,'PERMISSION'] = df_permission[(df_permission['YEAR']==year) & (df_permission['MONTH']==month)].PERMISSION.values[0]
        df.loc[i,'PERMISSION_RATE'] = df_permission[(df_permission['YEAR']==year) & (df_permission['MONTH']==month)].PERMISSION_RATE.values[0]

    df = df[['SOLDDATE', 'CUSTID', 'PRODNAME', 'REGION_U', 'REGION', 'TEMP', 'HUM', 'CONSTRUCTION', 'CONSTRUCTION_RATE','PERMISSION', 'PERMISSION_RATE', 'QUANT']]

    return df

df_real = concat_all(df_21,weather_data)
df_total = concat_all(date_21,weather_data)
```

### 고객번호와 상관없이 예측

```python
from xgboost import XGBRegressor
from sklearn.preprocessing import StandardScaler

# 전처리
scaler = StandardScaler()

y = df_total['QUANT']
X = df_total.drop(['SOLDDATE','CUSTID','REGION_U','REGION','PERMISSION','PERMISSION_RATE','QUANT'],axis=1)

for i in X.columns.tolist()[1:]:
    X[i] = scaler.fit_transform(X[i].values.reshape(-1,1))
X['QUANT'] = y

X_train = X[X.QUANT.notnull()].drop(['QUANT'],axis=1)
y_train = X[X.QUANT.notnull()]['QUANT']
X_test = X[X.QUANT.isnull()].drop(['QUANT'],axis=1)

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# 모델
model = XGBRegressor(n_estimators=1000, learning_rate=0.01, max_depth=5, n_jobs=-1)
model.fit(X_train, y_train)
pred = model.predict(X_test)
result = df_total.copy()
result.loc[result.QUANT.isnull() == True,'QUANT'] = pred
result['QUANT'] = result['QUANT'].apply(lambda x: round(x,-2))
```
