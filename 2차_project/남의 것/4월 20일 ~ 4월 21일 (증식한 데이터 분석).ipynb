# 고객별 구매 비율 --> 데이터 증식시 비율을 반영

```python
df2019 = df1.loc[(df1.SOLDDATE <= '2019-02-22') & (df1.SOLDDATE <= '2020-04-21')]
df2020 = df1.loc[(df1.SOLDDATE >= '2020-02-22') & (df1.SOLDDATE <= '2020-04-21')]
df2021 = df1.loc[(df1.SOLDDATE >= '2021-02-22') & (df1.SOLDDATE <= '2021-04-21')]
```

```python
df2019_cust=df2019.groupby(['CUSTID','PRODNAME'])[['QUANT']].sum().reset_index()
df2020_cust=df2020.groupby(['CUSTID','PRODNAME'])[['QUANT']].sum().reset_index()
df2021_cust=df2021.groupby(['CUSTID','PRODNAME'])[['QUANT']].sum().reset_index()
```

```python
for cust in df2019_cust['CUSTID'].unique():
  df2019_cust.loc[df2019_cust['CUSTID']==cust,'QUANT_RATE'] = df2019_cust.loc[df2019_cust['CUSTID']==cust]['QUANT'] / df2019_cust.loc[df2019_cust['CUSTID']==cust]['QUANT'].sum()

for cust in df2020_cust['CUSTID'].unique():
  df2020_cust.loc[df2020_cust['CUSTID']==cust,'QUANT_RATE'] = df2020_cust.loc[df2020_cust['CUSTID']==cust]['QUANT'] / df2020_cust.loc[df2020_cust['CUSTID']==cust]['QUANT'].sum()

for cust in df2021_cust['CUSTID'].unique():
  df2021_cust.loc[df2021_cust['CUSTID']==cust,'QUANT_RATE'] = df2021_cust.loc[df2021_cust['CUSTID']==cust]['QUANT'] / df2021_cust.loc[df2021_cust['CUSTID']==cust]['QUANT'].sum()
```

```python
merge_outer = pd.merge(df2019_cust,df2020_cust, how='outer',on=['CUSTID','PRODNAME'])
```

```python
merge_outer = pd.merge(merge_outer,df2021_cust, how='outer',on=['CUSTID','PRODNAME'])
```

```python
merge_outer.columns=['CUSTID','PRODNAME','19_QUANT','19_RATE','20_QUANT','20_RATE','21_QUANT','21_RATE']
```

```python
merge_outer=merge_outer.sort_values(['CUSTID','PRODNAME'])
```

```python
merge_outer['idx']=merge_outer['CUSTID']+' - '+merge_outer['PRODNAME']
merge_outer=merge_outer.drop(['CUSTID', 'PRODNAME'], axis=1)
merge_outer=merge_outer.set_index('idx')
```

```python
merge_outer=merge_outer.fillna(0)
```
