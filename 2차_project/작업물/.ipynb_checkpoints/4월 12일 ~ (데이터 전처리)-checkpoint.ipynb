```python
import pymysql
import pandas as pd
import numpy as np
import re
```

```python
connection = pymysql.connect(host='127.0.0.1', user='root', password='root', db='smart_factory', charset='utf8')   # 접속정보
cursor = connection.cursor()  
sql = "SELECT 제품코드, 제품명 FROM `product`"   
cursor.execute(sql)  
result=cursor.fetchall()
columns = ['제품코드', '제품명']
df = pd.DataFrame(result, columns = columns)
connection.commit()
connection.close()
```

```python
df
```

# # product 테이블에서 제품명 컬럼을 제품과 지역으로 분리
---

```python
products = []
for i in df['제품명']:
    a = re.search('[a-zA-Z]+-*[0-9a-zA-Z]+', i).group()
    products.append(a)
    
products
```

```python
regions = []
for i in df['제품명']:
    a = re.search(r'\([가-힣]*\)', i)
    if a == None:
        a = re.search('[가-힣]+', i)
        regions.append(a.group())
        
    else:
        b = re.sub(r'\(*\)*', '', a.group())
        regions.append(b)

regions
```

```python
df['제품'] = products
df['지역'] = regions
```

```python
df_re = df.drop('제품명', axis = 1)

t = []
for i in range(len(df_re)):
    data = tuple(df_re.loc[i])
    t.append(data)
```

# # 원자재 중량을 원자재 비율로 변경
---

```python
conn = pymysql.connect(host='127.0.0.1', user='root', password='root', db='smart_factory')
sql_state = 'SELECT * FROM `2020_raw`'
df_2020 = pd.read_sql_query(sql_state, conn)
display(df_2020.head())
```

```python
# 숫자로 변경

SR35761 = []
for i in df_2020['SR35761']:
    SR35761.append(float(re.sub('\r', '', i)))
SR35761

df_2020['SR35761'] = SR35761
```

```python
# 컬럼명 변경

material = df_2020.columns[4:]
recol = list(df_2020.columns[:4])
for i in df_2020.columns[4:]:
    rename = i + '_weight'
    recol.append(rename)
df_2020.columns = recol
display(df_2020.head())
```

```python
# 비율 구하여 컬럼 추가

for i in list(material):
    df_2020[i] = round((df_2020[f'{i}_weight'] / df_2020['SOLD_QUANT']) * 100).astype(int)
    
df_2020_re = df_2020.drop(df_2020.columns[4:32], axis = 1)
a = df_2020_re.loc[df_2020_re['PRODNAME'] == 'CSA4000'].sort_values('SOLDDATE').iloc[0]
a[3:][a[3:] > 0]
```
