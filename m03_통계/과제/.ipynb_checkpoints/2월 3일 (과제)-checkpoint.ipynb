```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
plt.style.use('seaborn')
```

##### [과제] 포아송 분포의 표본평균 분포는 포아송 분포를 따르지 않고 좌우대칭인 정규분포에 근사한 형태가 되는 것을 입증하세요.
---

```python
lam = 3
rv = stats.poisson(lam)
n = 10

sample_size = int(1e6)
xs_sample = rv.rvs((n, sample_size))
sample_mean = np.mean(xs_sample, axis = 0)

np.mean(sample_mean), np.var(sample_mean, ddof = 1)
```

```python
rv1 = stats.norm(np.mean(sample_mean), np.sqrt(np.var(sample_mean, ddof = 1)))
xs = np.linspace(rv1.ppf(0.001), rv1.ppf(0.999), 100)

plt.figure(figsize = (10, 6))
plt.hist(sample_mean, bins = 100, density = True, label = '10 * Poi(3)')
plt.plot(xs, rv1.pdf(xs), label = 'N(sample_mean, sample_dev)', color = 'gold', linewidth = 3)

plt.legend(fontsize = 13)
plt.xlim(0, 6)
plt.show()
```

##### [과제] 중심극한정리를 입증하세요.
---

```python
lam = 3
rv = stats.poisson(lam)

n = 10000
sample_size = 10000
xs_sample = rv.rvs((n, sample_size))
sample_mean = np.mean(xs_sample, axis = 0)
```

```python
rv1 = stats.norm(lam, np.sqrt(lam/n))
xs = np.linspace(rv1.ppf(0.001), rv1.ppf(0.999), 100)

plt.figure(figsize = (10, 6))
plt.hist(sample_mean, bins = 100, density = True, label = '10000 * Poi(3)')
plt.plot(xs, rv1.pdf(xs), label = 'N(3, 3/10000)', color = 'gold', linewidth = 2)
plt.legend(fontsize = 16)
plt.show()
```

##### [과제] sample_size의 변화에 따라 표본평균이 모평균에 수렴함을 입증하세요. (대수의 법칙)
---

```python
p = 1/6
rv = stats.bernoulli(p)

n = int(1e5)
sample = rv.rvs((n, 4))
space = np.linspace(100, n, 50).astype(int)
plot_list = np.array([np.mean(sample[:sp], axis = 0) for sp in space]).T # .T = .transpose()
```

```python
linestyles = ['-', '--', '-.', ':']
plt.figure(figsize = (15, 8))

for pl, ls in zip(plot_list, linestyles):
    plt.plot(space, pl, ls = ls)
    
plt.hlines(p, -1, n, 'black', linewidth = 2)
plt.show()
```

##### [과제] 모평균과 모분산의 점추정 사례를 파이썬 코드로 수행하세요.
---

```python
df = pd.read_csv('D:\python\cakd5\m3_통계\data\ch4_scores400.csv')
display(df)

scores = np.array(df.score)
print('모평균, 모분산 :', np.mean(scores),'/', np.var(scores))
```

```python
np.random.seed(0)
samples = np.random.choice(scores, size = (10000, 20))
samples
```

```python
# 모평균의 점추정

for i in range(5):
    s_mean = np.mean(samples[i])
    print(f'{i+1}번째 표본평균 : {s_mean}')
print()
print('위 다섯 표본평균의 합 :', np.mean([np.mean(samples[i]) for i in range(5)]), '\n')

print(np.mean(samples, axis = 1).mean()) # 대수의 법칙
print(np.mean(np.random.choice(scores, int(1e6))))
```

```python
# 모분산의 점추정

for i in range(5):
    s_var = np.var(samples[i], ddof = 1).round(4)
    print(f'{i+1}번째 표본분산 : {s_var}')
print()
print('위 다섯 표본분산의 합 :', np.mean([np.var(samples[i], ddof = 1) for i in range(5)]), '\n')

sample_vars0 = np.var(samples, axis = 1)
print(np.mean(sample_vars0))

sample_vars1 = np.var(samples, axis = 1, ddof = 1) # 불편분산
print(np.mean(sample_vars1))
```

```python
# 시험점수 데이터 불러오기
display(df)

# 모평균과 모분산 
p_mean = np.mean(scores)
p_var = np.var(scores)
print('모평균, 모분산:', p_mean, p_var)

# 시험점수 모집단(전교생 시험점수)이 정규분포를 따른다고 가정 
plt.figure(figsize=(10,6))
xs = np.arange(101)
rv = stats.norm(p_mean, np.sqrt(p_var))
plt.plot(xs, rv.pdf(xs), color='gray')
plt.hist(scores, bins=100, range=(0,100), density=True) # 전체 시험점수 히스토그램
plt.show()

# 무작위추출한 20명의 시험점수 표본 데이터
np.random.seed(0)
n = 20
sample = np.random.choice(scores, n)
print('시험점수 표본 데이터:', sample)

# 표본 크기가 20인 표본 데이터 1만 개 추출
# (표본으로 평균과 분산을 어느 정도의 정확도로 추출할 수 있는지 시뮬레이션)
np.random.seed(1111)
n_samples = 10000
samples = np.random.choice(scores, (n_samples, n))

## 대수의 법칙을 사용하여 기댓값이 모평균인 것을 확인
sample_means = np.mean(samples, axis=1) # 준비해둔 1만 개의 표본 데이터(n=20) 각각의 표본평균을 구하고
s_means = np.mean(sample_means)         # 그 표본평균들의 평균(기댓값) 계산
print('모평균:', p_mean, '\n', 
      '표본평균의 기댓값:', s_means)
```

```python
## 표본분산이 모분산의 불편추정량이 되는지 대수의 법칙으로 확인
sample_vars = np.var(samples, axis=1) # 준비해둔 1만 개의 표본 데이터 각각의 표본분산 구하고
s_vars = np.mean(sample_vars)         # 표본분산의 평균 
## 불편분산이 모분산의 불편추정량이 되는지 확인
sample_u_vars = np.var(samples, axis=1, ddof=1)
u_vars = np.mean(sample_u_vars)
print('모분산:', p_var, '\n', 
      '표본분산:', s_vars, '\n', 
     '불편분산:', u_vars)
```

##### [과제] 모분산을 알지 못할 때 정규분포의 모평균의 구간 추정을 파이썬 코드로 수행하세요.
---

모분산을 모르는 경우
- $\sqrt{\sigma^2/n}$ 대신 불편표준편차의 $\sqrt{s^2/n}$를 표준오차로 사용한다.
- 표준화 과정에서 다음의 식이 사용되는데, 이것은 t분포를 따른다. (카이제곱분포 Y 이용하여 구함)  
$t = $ $Z \over \sqrt{Y/(n-1)}$
- 이 때, $\alpha$ 신뢰구간은 다음과 같다.  
$[\bar{X}-t_{\alpha/2}(n-1)\sqrt{s^2/n}, \bar{X}-t_{1-\alpha/2}(n-1)\sqrt{s^2/n}]$


```python
print('모평균, 모분산 :', np.mean(scores),'/', np.var(scores), '\n')
u_var = np.var(samples[1], ddof = 1)
n = 20

rv = stats.t(n - 1)
lcl = np.mean(scores) - rv.ppf(0.975) * np.sqrt(u_var/n) # 신뢰하한
ucl = np.mean(scores) - rv.ppf(0.025) * np.sqrt(u_var/n) # 신뢰상한

lcl, ucl # 모평균의 95% 신뢰구간
```

##### [과제] 모집단에 정규분포를 가정할 수 없는 예를 포아송 분포로 설정하여 모평균 신뢰구간을 구하세요
---

```python
df = pd.read_csv('D:\python\cakd5\m3_통계\data\ch10_access.csv')
display(df)
n_access = np.array(df['access number'])
n = len(n_access)

s_mean = n_access.mean().round(3) # 포아송 분포 표본평균 분포의 기대값 / # 포아송 분포 표본평균 분포의 분산 : 기대값/n
s_mean
```

```python
rv = stats.norm()
lcl = s_mean - rv.isf(0.025) * np.sqrt(s_mean/n)
ucl = s_mean - rv.isf(0.975) * np.sqrt(s_mean/n)

lcl, ucl
```
