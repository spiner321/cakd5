- 2Í∞ú Ïù¥ÏÉÅÏùò ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÑúÎ°ú ÏòÅÌñ•ÏùÑ ÎÅºÏπòÏßÄ ÏïäÏúºÎ©∞ Í¥ÄÍ≤åÍ∞Ä ÏóÜÏùåÏùÑ ÎÇòÌÉÄÎÇ¥Îäî Í∞úÎÖê

 - $f_{X, Y}(x,y) = f_X(x)f_Y(y)$
- ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÎèÖÎ¶ΩÏùº Îïå Í≤∞Ìï©ÌôïÎ•†ÏùÄ Ï£ºÎ≥ÄÌôïÎ•†Ïùò Í≥±ÏúºÎ°ú ÏûëÏÑ±Ìï† Ïàò ÏûàÎã§.

 - $f_{X_1, X_2, ..., X_n}(x_1, x_2, ..., x_n) = f_{X-1}(x_1)f_{X_2}(x_2)...f_{X_n}(x_n)$
- Ìï®Ïàò fÍ∞Ä Ïù¥ÏÇ∞ÌòïÏù¥Î©¥ ÌôïÎ•†Ìï®Ïàò, Ïó∞ÏÜçÌòïÏù¥Î©¥ Î∞ÄÎèÑÌï®ÏàòÎ•º ÎÇòÌÉÄÎÇ∏Îã§.

#### # ÎèÖÎ¶ΩÏÑ±Í≥º Î¨¥ÏÉÅÍ¥ÄÏÑ±
---
- Í≥µÎ∂ÑÏÇ∞Ïù¥ÎÇò ÏÉÅÍ¥ÄÍ≥ÑÏàòÍ∞Ä 0Ïùº Îïå Î¨¥ÏÉÅÍ¥ÄÏù¥Îùº ÌïòÍ≥† Îëê ÌôïÎ•†Î≥ÄÏàò ÏÇ¨Ïù¥ÏóêÎäî ÏÉÅÍ¥ÄÏÑ±, Ï¶â ÏÑ†ÌòïÍ¥ÄÍ≥ÑÍ∞Ä ÏóÜÏùåÏùÑ ÎÇòÌÉÄÎÉÑ
- ÎèÖÎ¶ΩÏÑ±ÎèÑ Î¨¥ÏÉÅÍ¥ÄÏÑ±ÎèÑ Îëê ÌôïÎ•†Î≥ÄÏàòÍ∞Ä ÏÑúÎ°ú Í¥ÄÍ≥Ñ ÏóÜÎã§Îäî Í≤ÉÏùÑ ÎÇòÌÉÄÎÇ¥Îäî ÏÑ±ÏßàÏù¥Îã§.
 - Î¨¥ÏÉÅÍ¥ÄÏÑ±Î≥¥Îã§ ÎèÖÎ¶ΩÏÑ±Ïù¥ Îçî Í∞ïÌïú Í∞úÎÖê
 - 2Í∞úÏùò ÌôïÎ•†Î≥ÄÏàò XÏôÄ YÍ∞Ä ÎèÖÎ¶ΩÏùº Îïå XÏôÄ YÎäî Î¨¥ÏÉÅÍ¥ÄÏù¥ ÎêòÏßÄÎßå XÏôÄ YÍ∞Ä Î¨¥ÏÉÅÍ¥ÄÏùº Îïå XÏôÄ YÍ∞Ä Î∞òÎìúÏãú ÎèÖÎ¶ΩÏù∏ Í≤ÉÏùÄ ÏïÑÎãàÎã§.

```python
# ÏÇ¨Ïö©Ïûê Ï†ïÏùò Ìï®Ïàò

def f_XY(x, y):
    if x in x_set and y in y_set:
        return x * y / 441
    else:
        return 0
XY = [x_set, y_set, f_XY]

def E(XY, g):
    x_set, y_set, f_XY = XY
    return np.sum([g(i, j) * f_XY(i, j) for i in x_set for j in y_set])

def Cov(XY):
    x_set, y_set, f_XY = XY
    mean_X = E(XY, lambda x, y: x)
    mean_Y = E(XY, lambda x, y: y)
    return np.sum([(i - mean_X) * (j - mean_Y) * f_XY(i, j) for i in x_set for j in y_set])

def f_X(x):
    return np.sum([f_XY(x, y_k) for y_k in y_set])

def f_Y(y):
    return np.sum([f_XY(x_k, y) for x_k in x_set])
```

### # Ìï©Ïùò Î∂ÑÌè¨
---
- ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÌôïÎ•£ÎÖÄÏàò $X_1, X_2, ..., X_nÏùò Ìï© \sum_{i=1}^n X_i = X_1 + X_2 + ... + X_n$Ïù¥ Îî∞Î•¥Îäî ÌôïÎ•†Î∂ÑÌè¨Î•º ÏùòÎØ∏
- Ìï©Ïùò Î∂ÑÌè¨Î•º Ïù¥Ìï¥ÌïòÎ©¥ ÌôïÎ•†Î≥ÄÏàòÏùò Ìï©ÏùÑ ÎÇòÎàà ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨ÎèÑ ÏâΩÍ≤å Ïù¥Ìï¥Ìï† Ïàò ÏûàÎã§.
- Ïû¨ÏÉùÏÑ± : ÎèôÏùºÌïú Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî 2Í∞úÏùò ÎèÖÎ¶ΩÏù∏ ÌôïÎ•†Î≥ÄÏàòÏóê ÎåÄÌï¥ÏÑú, Í∑∏ Ìï©ÎèÑ ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Í∞Ä ÎêòÎäî ÏÑ±Ïßà

#### # Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
---

```python
# ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ X ~ N(1, 2), Y ~ N(2, 3)Ïù¥ ÏûàÏùÑ Îïå ÌôïÎ•†Î≥ÄÏàò X + YÏùò Î∂ÑÌè¨Îäî N(3, 5)Î•º Îî∞Î•∏Îã§.
np.random.seed(0)

rv1 = stats.norm(1, np.sqrt(2))
rv2 = stats.norm(2, np.sqrt(3))
rv3 = stats.norm(3, np.sqrt(5))

sample_size = int(1e6)
X_sample = rv1.rvs(sample_size)
Y_sample = rv2.rvs(sample_size)
XY_sample = rv3.rvs(sample_size)
sum_sample = X_sample + Y_sample

print(f'ÌèâÍ∑† : {np.mean(X_sample), np.mean(Y_sample), np.mean(XY_sample)}')
print(f'Î∂ÑÏÇ∞ : {np.var(X_sample), np.var(Y_sample), np.var(XY_sample)}')
print(f'Ìï©Ïùò ÌèâÍ∑†, Î∂ÑÏÇ∞ : {np.mean(sum_sample), np.var(sum_sample)}')
```

```python
plt.figure(figsize = (10, 6))
rv = stats.norm(3, np.sqrt(5))
xs = np.linspace(rv.isf(0.995), rv.isf(0.005), 100)

plt.hist(sum_sample, bins = 100, density = True, alpha = 0.5, label = 'N(1, 2) + N(2, 3)')
plt.plot(xs, rv.pdf(xs), label ='N(3, 5)', color = 'gray')
plt.plot(xs, rv1.pdf(xs), label = 'N(1, 2)', ls = '--', color = 'gray')
plt.plot(xs, rv2.pdf(xs), label = 'N(2, 3)', ls = ':', color = 'gray')

plt.legend(fontsize = 13)
plt.xlim(rv.isf(0.995), rv.isf(0.005))
```

#### # Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
---
- ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ X ~ Poi(3)Í≥º Y ~ Poi(4) Í∞Ä ÏûàÏùÑ Îïå ÌôïÎ•†Î≥ÄÏàò X + YÎèÑ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Î•º Îî∞Î•¥Í≥† X + Y ~ Poi(7)Ïù¥ ÎêúÎã§.
- nÍ∞úÏùò Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò Ìï©Ïù¥ ÎêòÎçîÎùºÎèÑ Ïû¨ÏÉùÏÑ±Ïóê ÏùòÌïòÏó¨ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Î•º Îî∞Î•∏Îã§.

```python
rv1 = stats.poisson(3)
rv2 = stats.poisson(4)
rv3 = stats.poisson(7)

sample_size = int(1e6)
X_sample = rv1.rvs(sample_size)
Y_sample = rv2.rvs(sample_size)
XY_sample = rv3.rvs(sample_size)
sum_sample = X_sample + Y_sample

print(f'ÌèâÍ∑† : {np.mean(X_sample), np.mean(Y_sample), np.mean(XY_sample)}')
print(f'Î∂ÑÏÇ∞ : {np.var(X_sample), np.var(Y_sample), np.var(XY_sample)}')
print(f'Ìï©Ïùò ÌèâÍ∑†, Î∂ÑÏÇ∞ : {np.mean(sum_sample), np.var(sum_sample)}')
```

```python
# ÏÑúÎ°ú ÎèÖÎ¶ΩÏù∏ nÍ∞úÏùò ÌôïÎ•†Î≥ÄÏàòÍ∞Ä Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïóê ÎåÄÌïòÏó¨ Í∑∏ Ìï©ÎèÑ Ïû¨ÏÉùÏÑ±Ïóê ÏùòÌï¥ Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨Ïùò ÌòïÌÉúÎ•º Ïú†ÏßÄÌïúÎã§.

plt.figure(figsize = (10, 6))
rv = stats.poisson(7)
xs = np.arange(20)
hist, _ = np.histogram(sum_sample, bins = 20, range = (0, 20), normed = True) # normed = TrueÎäî Ï†ïÍ∑úÌôîÎ•º ÌïòÏó¨ 
plt.bar(xs, hist, alpha = 0.5, label = 'Poi(3) + Poi(4)')
plt.plot(xs, rv.pmf(xs), label = 'Poi(7)', color = 'gray')
plt.plot(xs, rv1.pmf(xs), label = 'Poi(3)', ls = '--', color = 'gray')
plt.plot(xs, rv2.pmf(xs), label = 'Poi(4)', ls = ':', color = 'gray')

plt.legend(fontsize = 13)
```

#### # Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïùò Ìï©Ïùò Î∂ÑÌè¨
---
- Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ÏóêÎäî Ïû¨ÏÉùÏÑ±Ïù¥ ÏóÜÎã§.
- Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïùò Ìï©ÏùÄ Ïù¥Ìï≠Î∂ÑÌè¨Í∞Ä ÎêòÎäî ÏÑ±ÏßàÏù¥ ÏûàÎã§.

```python
# ÏÑ±Í≥µ ÌôïÎ•†Ïù¥ 0.3Ïù∏ Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÌôïÎ•†Î≥ÄÏàò 10Í∞úÏóê ÎåÄÌïòÏó¨ Î∞±ÎßåÎ≤à ÌëúÎ≥∏ÏùÑ Ï∂îÏ∂úÌïòÏó¨ ÌëúÎ≥∏ÌèâÍ∑†Í≥º ÌëúÎ≥∏Î∂ÑÏÇ∞ÏùÑ ÏÇ∞Ï∂ú
# ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Í∞Ä ÌïòÎÇòÏù¥Î©¥ Î≤†Î•¥ÎàÑÏù¥Î∂ÑÌè¨Í∞Ä ÎêòÍ≥†, ÌëúÎ≥∏ Îç∞Ïù¥ÌÑ∞Í∞Ä Ïó¨ÎüøÏù¥Î©¥ Ïù¥Ìï≠Î∂ÑÌè¨Í∞Ä ÎêúÎã§.

np.random.seed(0)

p = 0.3
rv = stats.bernoulli(p)
sample_size = int(1e6)
X_sample = rv.rvs((10, sample_size))
print('ÌëúÎ≥∏ Ï∂îÏ∂ú :' , X_sample.shape, '\n', 'Ï≤´Î≤àÏû¨ Ïó¥ :', X_sample[:, 0], '\n')
sum_sample = np.sum(X_sample, axis = 0)
print('ÏÑ±Í≥µ ÌöüÏàò :', sum_sample, '\n')
print('ÏÑ±Í≥µ ÌöüÏàòÏùò ÌèâÍ∑† : ', np.mean(sum_sample),'\nÏÑ±Í≥µ ÌöüÏàòÏùò Î∂ÑÏÇ∞ : ', np.var(sum_sample, ddof = 1))
```

```python
fig = plt.figure(figsize = (10, 6))
ax = fig.add_subplot(111)

rv = stats.binom(10, p)
xs = np.arange(10)
hist, _ = np.histogram(sum_sample, bins = 10, range = (0, 10), density = True)
plt.bar(xs, hist, alpha = 0.5, label = 'sum of 10 * bern(0.3)')
plt.plot(xs, rv.pmf(xs), label = 'Bin(10, 0.3)')
plt.legend(fontsize = 13)
plt.xlim(-0.5, 10)
plt.xticks(np.arange(10))
plt.show()
```

### # ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨
---
- ÌëúÎ≥∏ÌèâÍ∑†Ïùò Î∂ÑÌè¨ÎûÄ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† ÎèôÏùºÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•¥Îäî ÌôïÎ•†Î≥ÄÏàò $X_1, X_2, ..., X_n$Ïùò ÌëúÎ≥∏ÌèâÍ∑†Ïù¥ Îî∞Î•¥Îäî Î∂ÑÌè¨
- Ïù¥ Î∂ÑÌè¨Îäî Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨ÎÇò ÎåÄÏàòÏùò Î≤ïÏπô Îì± Ï∂îÏ∏°ÌÜµÍ≥ÑÏóêÏÑú Ï§ëÏöîÌïú ÏÑ±ÏßàÏùÑ Îã¥Í≥† ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê Ïú†ÏùòÌï¥Ïïº ÌïúÎã§.
- $E(\bar{X}) = \mu$  
- $V(\bar{X}) = {\sigma}^2/n)$

#### # Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌëúÎ≥∏ÌèâÍ∑† Î∂ÑÌè¨
---

```python
mean = 1
var = 2
rv = stats.norm(mean, np.sqrt(var))

n = 10
sample_size = int(1e6)
X_sample = rv.rvs((n, sample_size))
print('ÌëúÎ≥∏ ÏÇ¨Ïù¥Ï¶à 10, ÌëúÎ≥∏Í∞úÏàò 1000000 :', X_sample.shape, '\n')
sample_mean = np.mean(X_sample, axis = 0)
print('ÌëúÎ≥∏ÌèâÍ∑† Î∞±ÎßåÍ∞ú :', sample_mean, len(sample_mean), '\n')
print('ÌëúÎ≥∏ÌèâÍ∑† Î∞±ÎßåÍ∞úÏùò ÌèâÍ∑† :', np.mean(sample_mean),'\nÌëúÎ≥∏ÌèâÍ∑† Î∞±ÎßåÍ∞úÏùò Î∂ÑÏÇ∞ :', np.var(sample_mean))
```

```python
plt.figure(figsize = (10, 6))
rv_true = stats.norm(mean, np.sqrt(var/n))
xs = np.linspace(rv_true.ppf(0.001), rv_true.ppf(0.999), 100)
plt.hist(sample_mean, bins = 100, density = True, alpha = 0.5, label = 'sample mean of 10 * N(1, 2)')
plt.plot(xs, rv_true.pdf(xs), label = 'N(1, 0.2)', color = 'gray')
plt.legend(fontsize = 13)
plt.xlim(rv_true.ppf(0.001), rv_true.ppf(0.999))
plt.show()
```

### # Ï§ëÏã¨Í∑πÌïúÏ†ïÎ¶¨
---
ÌôïÎ•†Î≥ÄÏàò $X_1, X_2, ..., X_n$Ïù¥ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥† Í∏∞ÎåÄÍ∞íÏù¥ $\mu$, Î∂ÑÏÇ∞Ïù¥ ${\sigma}^2/n$Ïù∏ ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•º Îïå nÏù¥ Ïª§ÏßêÏóê Îî∞Îùº ÌëúÎ≥∏ÌèâÍ∑† $\bar{X}$Ïùò Î∂ÑÌè¨Îäî Ï†ïÍ∑úÎ∂ÑÌè¨ $N(\mu, \sigma^2/n)$Ïóê Í∞ÄÍπåÏõåÏßÑÎã§.

### # ÎåÄÏàòÏùò Î≤ïÏπô
---
- ÌôïÎ•†Î≥ÄÏàò $X_1, X_2, ..., X_n$Ïù¥ ÏÑúÎ°ú ÎèÖÎ¶ΩÏù¥Í≥†ÌèâÍ∑†Ïù¥ $\mu$, Î∂ÑÏÇ∞Ïù¥ ${\sigma}^2$Ïùº Í≤É Í∞ôÏùÄ ÌôïÎ•†Î∂ÑÌè¨Î•º Îî∞Î•º Îïå nÏù¥ Ïª§ÏßêÏóê Îî∞Îùº ÌëúÎ≥∏ÌèâÍ∑†ÏùÄ $\mu$Ïóê ÏàòÎ†¥ÌïúÎã§.
